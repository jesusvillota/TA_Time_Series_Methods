<h2 class="styled-header">11) Non-stationarity & Integration</h2>

### **11.1 Integration Order: I(0) vs I(1)**

A time series $Y_t$ is said to be **integrated of order $d$**, denoted $Y_t \sim I(d)$, if it requires $d$ differences to become stationary.

**Key Definitions:**

- **I(0)**: Stationary series (no differencing needed)
  - Mean and variance constant over time
  - Autocovariance depends only on lag, not time
  
- **I(1)**: Series that becomes stationary after **first differencing**
  - $Y_t \sim I(1)$ if $\Delta Y_t = Y_t - Y_{t-1} \sim I(0)$
  - Most common in economics: prices, exchange rates, GDP levels
  
- **I(d)**: General case requiring $d$ differences

**Economic Interpretation:**
- **I(0)**: Returns, growth rates, inflation (fluctuate around a constant mean)
- **I(1)**: Prices, exchange rates, GDP levels (wander without bound)

---

### **11.2 Random Walk Models**

The simplest **I(1)** process is the **random walk**:

**Random Walk (No Drift):**
$$Y_t = Y_{t-1} + \varepsilon_t, \quad \varepsilon_t \sim \text{i.i.d.}(0, \sigma^2)$$

**Properties:**
- $Y_t = Y_0 + \sum_{i=1}^{t} \varepsilon_i$ (cumulative sum of shocks)
- $\mathbb{E}[Y_t] = Y_0$ (constant mean)
- $\text{Var}(Y_t) = t\sigma^2$ ✗ **Non-stationary** (variance grows with time!)
- $\Delta Y_t = \varepsilon_t \sim I(0)$ ✓ **First difference is stationary**

**Random Walk with Drift:**
$$Y_t = \mu + Y_{t-1} + \varepsilon_t$$

**Properties:**
- $Y_t = Y_0 + \mu t + \sum_{i=1}^{t} \varepsilon_i$
- Mean: $\mathbb{E}[Y_t] = Y_0 + \mu t$ ✗ **Non-stationary** (trending mean)
- Variance: $\text{Var}(Y_t) = t\sigma^2$ ✗ **Non-stationary**
- $\Delta Y_t = \mu + \varepsilon_t \sim I(0)$ ✓ **First difference is stationary around drift**

---

### **11.3 Deterministic vs Stochastic Trends**

**Deterministic Trend:**
$$Y_t = \beta_0 + \beta_1 t + \varepsilon_t, \quad \varepsilon_t \sim I(0)$$

- Trend is **fixed** (deterministic function of time)
- Removing trend: $Y_t - \beta_1 t = \beta_0 + \varepsilon_t$ is stationary
- **Trend-stationary**: I(0) after detrending

**Stochastic Trend (Unit Root):**
$$Y_t = Y_{t-1} + \varepsilon_t = Y_0 + \sum_{i=1}^{t} \varepsilon_i$$

- Trend is **random** (cumulative sum of shocks)
- Removing deterministic trend **does not** make it stationary
- **Difference-stationary**: I(1), requires differencing

**Key Distinction:**
- **Deterministic trend**: Detrend → stationary
- **Stochastic trend**: Difference → stationary

**Why It Matters:**
- **Forecasting**: Deterministic trends predictable; stochastic trends not
- **Spurious regression**: Two I(1) series can appear correlated even if independent
- **Cointegration**: Two I(1) series may share a common stochastic trend

---

### **11.4 Visual Intuition: EUR/USD Exchange Rate**

The plots above show the EUR/USD exchange rate in levels and first differences. Notice how:
- The **level** (`lneurusd`) appears to wander without bound - a classic sign of non-stationarity
- The **first difference** (`d_lneurusd`) fluctuates around zero - suggesting stationarity

This visual inspection provides initial evidence that the exchange rate level may be I(1), while returns (first differences) are I(0). We will now test this formally using unit root tests.

<h2 class="styled-header">12) Dickey-Fuller Tests</h2>

### **12.1 The Dickey-Fuller (DF) Test**

The **Dickey-Fuller test** is the most common test for unit roots. It tests the null hypothesis that a series has a unit root (is I(1)) against the alternative that it is stationary (I(0)).

**Basic DF Regression:**

Consider an AR(1) process: $Y_t = \phi Y_{t-1} + \varepsilon_t$

**Null hypothesis** ($H_0$): $\phi = 1$ (unit root, non-stationary)

**Alternative hypothesis** ($H_1$): $|\phi| < 1$ (stationary)

**Test Regression:**
$$\Delta Y_t = (\phi - 1) Y_{t-1} + \varepsilon_t = \delta Y_{t-1} + \varepsilon_t$$

where $\delta = \phi - 1$.

**Test Statistic:**
$$t_{\delta} = \frac{\hat{\delta}}{se(\hat{\delta})}$$

**Critical Values:**
- ✗ **Non-standard distribution** (not Student's $t$!)
- ✗ Critical values are **more negative** than standard $t$-test
- ✗ Must use **Dickey-Fuller critical values** (or MacKinnon approximate $p$-values)

**Decision Rule:**
- If $t_{\delta} < \text{DF critical value}$: **Reject $H_0$** → Series is **stationary** (I(0))
- If $t_{\delta} > \text{DF critical value}$: **Fail to reject $H_0$** → Series has **unit root** (I(1))

---

### **12.2 Three Forms of the DF Test**

The DF test has three variants depending on deterministic components:

**1. No Constant, No Trend:**
$$\Delta Y_t = \delta Y_{t-1} + \varepsilon_t$$

- Use when: Series has zero mean and no trend
- Rare in practice

**2. With Constant (Drift):**
$$\Delta Y_t = \alpha + \delta Y_{t-1} + \varepsilon_t$$

- Use when: Series may have non-zero mean but no trend
- Most common for **levels** of economic variables

**3. With Constant and Trend:**
$$\Delta Y_t = \alpha + \beta t + \delta Y_{t-1} + \varepsilon_t$$

- Use when: Series may have deterministic trend
- Use when: Visual inspection suggests trending behavior

**Which Form to Use?**
- **Start with constant + trend** if series appears to trend
- **Use constant only** if no clear trend
- **Check sensitivity**: Results should be robust to specification

---

### **12.3 Augmented Dickey-Fuller (ADF) Test**

The **Augmented Dickey-Fuller (ADF)** test extends DF to handle **serial correlation** in $\varepsilon_t$.

**ADF Regression:**
$$\Delta Y_t = \alpha + \beta t + \delta Y_{t-1} + \sum_{i=1}^{p} \gamma_i \Delta Y_{t-i} + \varepsilon_t$$

**Key Addition:**
- **Lagged differences** $\Delta Y_{t-1}, \ldots, \Delta Y_{t-p}$ control for serial correlation
- Ensures $\varepsilon_t$ is white noise (required for valid inference)

**Lag Selection:**
- **Too few lags**: Residual autocorrelation → biased test
- **Too many lags**: Lower power (harder to reject $H_0$)

**Common Rules:**
1. **Schwert (1989)**: $p_{\max} = \lfloor 12(T/100)^{1/4} \rfloor$
2. **Information Criteria**: Choose $p$ that minimizes AIC or BIC
3. **Sequential Testing**: Start with $p_{\max}$, drop lags if insignificant

**Stata Default:**
- `dfuller` uses **Schwert rule** by default
- Can specify `lags(#)` manually

---

### **12.4 Interpreting ADF Test Results**

**Test Output Contains:**
- **Test statistic** ($t$-statistic on $\delta$)
- **Critical values** at 1%, 5%, 10% levels
- **MacKinnon approximate $p$-value** (preferred for inference)

**Decision:**
- **$p$-value < 0.05**: Reject $H_0$ → Series is **stationary** (I(0))
- **$p$-value ≥ 0.05**: Fail to reject $H_0$ → Series has **unit root** (I(1))

**Important Notes:**
- ADF has **low power** (hard to reject $H_0$ when true)
- **Failure to reject** does not prove unit root (may be low power)
- Always check **differenced series** to confirm I(1) classification

````stata
* ===================================================
* Lag Selection Using Information Criteria
* ===================================================
* Select optimal lag length for ADF test using AIC/BIC
* This is more formal than arbitrary lag choice

* Ensure data is loaded
use "../../data/processed/eurusd_data.dta", clear
tsset date

* Method: Estimate ADF regression manually for different lag lengths
* and compare information criteria

* Calculate maximum lag using Schwert rule
quietly count
local T = r(N)
local pmax = floor(12*((`T'/100)^(1/4)))
display "Maximum lag (Schwert rule): " `pmax'

* Store AIC and BIC for different lag lengths
matrix ic_results = J(`pmax'+1, 3, .)
matrix colnames ic_results = Lags AIC BIC

* Test lag lengths from 0 to pmax
forvalues p = 0/`pmax' {
    quietly {
        * Estimate ADF regression manually
        gen L_lneurusd = L.lneurusd
        gen D_lneurusd = D.lneurusd
        
        * Create lagged differences
        forvalues i = 1/`p' {
            gen LD`i' = L`i'.D_lneurusd
        }
        
        * Estimate regression with trend
        if `p' > 0 {
            regress D_lneurusd L_lneurusd LD* _trend
        }
        else {
            regress D_lneurusd L_lneurusd _trend
        }
        
        * Extract AIC and BIC
        estat ic
        matrix ic = r(S)
        matrix ic_results[`p'+1, 1] = `p'
        matrix ic_results[`p'+1, 2] = ic[1,5]  // AIC
        matrix ic_results[`p'+1, 3] = ic[1,6]  // BIC
        
        * Clean up
        drop L_lneurusd D_lneurusd
        forvalues i = 1/`p' {
            capture drop LD`i'
        }
    }
}

* Display results
matrix list ic_results

* Find optimal lag (minimum AIC or BIC)
* Note: BIC is more conservative (prefers simpler models)
* For this demonstration, we'll use a simpler approach:

* Alternative simpler method: Use dfgls command if available
* or manually test a few lag lengths and compare

display "For practical purposes, test a few lag lengths:"
display "0, 4, 8, 12 lags and compare p-values"
display "Choose lag length where results are robust"
display "If p-values differ substantially, use information criteria above"
````

````stata
* ===================================================
* Residual Diagnostics After ADF Test
* ===================================================
* After running ADF test, check if residuals are white noise
* If residuals show autocorrelation, need more lags

* Ensure data is loaded
use "../../data/processed/eurusd_data.dta", clear
tsset date

* Run ADF test and store residuals
* First, estimate ADF regression manually to get residuals
gen L_lneurusd = L.lneurusd
gen D_lneurusd = D.lneurusd
gen LD1 = L1.D_lneurusd
gen LD2 = L2.D_lneurusd
gen LD3 = L3.D_lneurusd
gen LD4 = L4.D_lneurusd

* Estimate ADF(4) regression with trend
regress D_lneurusd L_lneurusd LD* _trend

* Store residuals
predict residuals, residuals

* Test residuals for autocorrelation
* Ljung-Box Q test
wntestq residuals, lags(10)

* Interpretation:
* - H0: No autocorrelation (white noise)
* - H1: Autocorrelation present
* - If p-value < 0.05: Reject H0 → Need more lags
* - If p-value >= 0.05: Fail to reject H0 → Residuals are white noise ✓

* Plot ACF of residuals
ac residuals, lags(20) title("ACF of ADF Residuals") ///
    note("Should show no significant autocorrelation if lag length is adequate")

* Clean up
drop L_lneurusd D_lneurusd LD* residuals

* Note: If residuals show autocorrelation, increase lag length and re-test
````

<h2 class="styled-header">13) Phillips-Perron Test</h2>

### **13.1 The Phillips-Perron (PP) Test**

The **Phillips-Perron test** is an alternative to ADF that uses **nonparametric corrections** for serial correlation and heteroskedasticity.

**Key Difference from ADF:**
- **ADF**: Adds **lagged differences** to control for serial correlation (parametric)
- **PP**: Uses **Newey-West correction** to adjust standard errors (nonparametric)

**PP Test Regression:**
$$\Delta Y_t = \alpha + \beta t + \delta Y_{t-1} + \varepsilon_t$$

**Same regression as DF**, but:
- **Standard errors** are adjusted using Newey-West HAC estimator
- **Test statistic** is modified to account for serial correlation
- **No lag selection needed** (nonparametric approach)

---

### **13.2 Advantages and Disadvantages**

**Advantages of PP:**
- ✓ **No lag selection** (avoids arbitrary choice of $p$)
- ✓ **Robust to heteroskedasticity** (uses HAC standard errors)
- ✓ **Same null/alternative** as ADF: $H_0$: unit root, $H_1$: stationary

**Disadvantages of PP:**
- ✗ **Bandwidth selection** needed for Newey-West (similar to lag selection)
- ✗ **Lower power** than ADF in some cases
- ✗ **Size distortions** in small samples

---

### **13.3 When to Use PP vs ADF?**

**Use ADF when:**
- Serial correlation structure is known/clear
- Want to model dynamics explicitly (lagged differences)
- Standard practice in most applications

**Use PP when:**
- Uncertain about lag structure
- Suspect heteroskedasticity
- Want robustness check on ADF results

**Best Practice:**
- **Report both tests** for robustness
- If **both reject** $H_0$: Strong evidence of stationarity
- If **both fail to reject**: Strong evidence of unit root
- If **conflicting results**: Investigate further (may indicate near-unit root or structural break)


````stata
* ===================================================
* Phillips-Perron Test: With Constant
* ===================================================
* Ensure data is loaded
use "../../data/processed/eurusd_data.dta", clear
tsset date

* Test log EUR/USD exchange rate using PP test
* PP uses nonparametric correction (no lag selection needed)
pperron lneurusd

* Interpretation:
* - Same null/alternative as ADF
* - H0: Unit root (non-stationary)
* - H1: Stationary
* - Compare p-value to ADF results for robustness
* - If p-value < 0.05: Reject H0 → Stationary (I(0))
* - If p-value >= 0.05: Fail to reject H0 → Unit root (I(1))
````

### **13.4 How to Read PP Test Output**

The Phillips-Perron test output is similar to ADF:

1. **Test Statistic (Z(t))**: Modified t-statistic with Newey-West correction
2. **Critical Values**: Same as ADF (non-standard distribution)
3. **MacKinnon p-value**: Use for inference

**Key Difference from ADF:**
- PP adjusts for serial correlation nonparametrically (no lag selection)
- Should give similar conclusions to ADF if both are correctly specified
- If results differ substantially, may indicate:
  - Heteroskedasticity (PP is more robust)
  - Near-unit root (different power properties)

**Interpretation:**
- p-value < 0.05 → Reject H0 → Stationary (I(0))
- p-value ≥ 0.05 → Fail to reject H0 → Unit root (I(1))

---

<h2 class="styled-header">14) KPSS Test</h2>

### **14.1 The KPSS Test: A Stationarity Test**

The **Kwiatkowski-Phillips-Schmidt-Shin (KPSS)** test is **complementary** to ADF/PP:

**Key Difference:**
- **ADF/PP**: $H_0$: Unit root (I(1)) vs $H_1$: Stationary (I(0))
- **KPSS**: $H_0$: Stationary (I(0)) vs $H_1$: Unit root (I(1))

**KPSS tests the null of stationarity** (opposite of ADF/PP)!

**Why Use KPSS?**
- **Confirmation**: If ADF fails to reject $H_0$ (unit root), KPSS can confirm by rejecting its $H_0$ (non-stationary)
- **Power issues**: ADF has low power; KPSS may detect non-stationarity when ADF doesn't
- **Complementary evidence**: Both tests together provide stronger inference

---

### **14.2 KPSS Test Forms**

**KPSS Test Regression:**
$$Y_t = \alpha + \beta t + r_t + \varepsilon_t$$

where $r_t = r_{t-1} + u_t$ is a random walk component.

**Null hypothesis** ($H_0$): $\text{Var}(u_t) = 0$ → $r_t$ is constant → $Y_t$ is stationary

**Alternative hypothesis** ($H_1$): $\text{Var}(u_t) > 0$ → $r_t$ is random walk → $Y_t$ has unit root

**Two Forms:**

1. **Level Stationary** (`notrend`):
   - $Y_t = \alpha + r_t + \varepsilon_t$ (no trend)
   - Tests stationarity around a constant mean

2. **Trend Stationary** (`trend`):
   - $Y_t = \alpha + \beta t + r_t + \varepsilon_t$ (with trend)
   - Tests stationarity around a deterministic trend

---

### **14.3 Interpreting KPSS Results**

**Test Statistic:**
- KPSS statistic (LM-type test)
- **Large values** → reject $H_0$ (non-stationary)
- **Small values** → fail to reject $H_0$ (stationary)

**Critical Values:**
- Provided at 1%, 5%, 10% levels
- **If test statistic > critical value**: Reject $H_0$ → Series is **non-stationary** (I(1))
- **If test statistic < critical value**: Fail to reject $H_0$ → Series is **stationary** (I(0))

**Decision Logic:**
- **ADF fails to reject** (unit root) **AND KPSS rejects** (non-stationary) → **Strong evidence of I(1)**
- **ADF rejects** (stationary) **AND KPSS fails to reject** (stationary) → **Strong evidence of I(0)**
- **Conflicting results** → Investigate further (near-unit root, structural break, or low power)

---

### **14.4 Caveats and Limitations**

**Bandwidth Selection:**
- KPSS requires **bandwidth parameter** for Newey-West correction
- Default in Stata: Automatic selection
- Can be sensitive to bandwidth choice

**Size Distortions:**
- KPSS can have **size problems** (reject $H_0$ too often) in small samples
- Use with caution for $T < 100$

**Complementary Use:**
- **Never use KPSS alone**; always combine with ADF/PP
- Use as **confirmation** of ADF/PP results


---

### **Decision Flowchart: Unit Root Testing Workflow**

```
START: Load data and plot series
  │
  ├─→ Visual inspection suggests trend?
  │     │
  │     YES → Use ADF/PP with constant + trend
  │     NO  → Use ADF/PP with constant only
  │
  ├─→ Run ADF test (with appropriate specification)
  ├─→ Run PP test (robustness check)
  ├─→ Run KPSS test (complementary evidence)
  │
  ├─→ Compare results using decision matrix
  │
  ├─→ Are results consistent?
  │     │
  │     YES → Test differenced series to confirm
  │     NO  → Investigate: structural breaks? near-unit root?
  │
  ├─→ Test differenced series
  │     │
  │     If level is I(1) → Difference should be I(0) ✓
  │
  └─→ Final decision: Use levels or differences for modeling
```

**Key Principle:** Always combine multiple tests and verify by testing differenced series!

---

**Important:** KPSS has the OPPOSITE null hypothesis from ADF/PP!

**KPSS Output Contains:**
1. **Test Statistic**: LM-type statistic (larger = more evidence of non-stationarity)
2. **Critical Values**: At 1%, 5%, 10% levels
3. **Decision Rule**: 
   - If test statistic > critical value → Reject H0 → Non-stationary (I(1))
   - If test statistic < critical value → Fail to reject H0 → Stationary (I(0))

**Combining with ADF/PP:**
- ADF fails to reject (unit root) AND KPSS rejects (non-stationary) → Strong evidence of I(1) ✓
- ADF rejects (stationary) AND KPSS fails to reject (stationary) → Strong evidence of I(0) ✓

### **16.1 Deterministic Component Selection**

**Guidelines:**
1. **Visual inspection**: Plot the series first
   - Clear upward/downward trend → Use **constant + trend**
   - Fluctuates around constant level → Use **constant only**
   - Appears to wander → Use **constant + trend** (most conservative)

2. **Economic theory**: 
   - Prices, exchange rates → Usually **constant + trend**
   - Returns, growth rates → Usually **constant only**

3. **Sensitivity check**: 
   - Test with both specifications
   - If results differ, report both and discuss

### **16.2 Lag Selection for ADF**

**Methods:**
1. **Schwert Rule** (default in Stata):
   $$p_{\max} = \lfloor 12(T/100)^{1/4} \rfloor$$
   - Automatic, data-driven
   - Good starting point

2. **Information Criteria**:
   - Estimate ADF with different $p$
   - Choose $p$ that minimizes **AIC** or **BIC**
   - More formal, but computationally intensive

3. **Sequential Testing**:
   - Start with $p_{\max}$
   - Drop lags if $t$-statistic on last lag < 1.96
   - Stop when all lags significant

4. **Residual Diagnostics**:
   - After ADF, test residuals for autocorrelation
   - If significant autocorrelation, increase $p$

**Best Practice:**
- Use **Schwert rule** as default
- Check **sensitivity** to lag choice
- Report results for multiple lag lengths if conclusions differ

---

### **16.3 Structural Breaks**

**Problem:**
- Unit root tests assume **constant parameters** over time
- **Structural breaks** (regime changes) can cause:
  - False rejection of $H_0$ (spurious stationarity)
  - False failure to reject $H_0$ (spurious unit root)

**Solutions:**
1. **Visual inspection**: Look for obvious breaks in plots
2. **Chow test**: Test for parameter stability
3. **Split sample**: Test pre- and post-break separately
4. **Advanced tests**: Zivot-Andrews test (allows for break under $H_0$)

**Common Breaks:**
- Financial crises (2008, 2020)
- Policy regime changes
- Currency devaluations

### **16.4 Finite-Sample Power**

**Power Problem:**
- Unit root tests have **low power** (hard to reject $H_0$ when false)
- Especially problematic for:
  - **Near-unit roots** ($\phi \approx 0.95$)
  - **Small samples** ($T < 100$)
  - **High persistence** (slow mean reversion)

**Implications:**
- **Failure to reject** $H_0$ does **not prove** unit root
- May be **low power** (true $\phi = 0.95$, but test can't detect)
- Always check **differenced series** to confirm I(1)

**Improving Power:**
- Use **longer samples** when possible
- Combine **multiple tests** (ADF + PP + KPSS)
- Use **panel unit root tests** if multiple series available

**Issumes:**
- **Outliers** can affect test statistics

- **Data errors** can create spurious patterns

**Pre-Testing Checks:**
1. **Summary statistics**: Check for extreme values

3. **Missing data**: Handle appropriately (interpolation vs. deletion)
4. **Data quality**: Verify source and transformations

**Robustness:**
- Test with and without outliers
- Compare results across different sample periods
- Use **robust standard errors** when appropriate

<h2 class="styled-header">11) Non-stationarity & Integration</h2>

### **11.1 Integration Order: I(0) vs I(1)**

A time series $Y_t$ is said to be **integrated of order $d$**, denoted $Y_t \sim I(d)$, if it requires $d$ differences to become stationary.

**Key Definitions:**

- **I(0)**: Stationary series (no differencing needed)
  - Mean and variance constant over time
  - Autocovariance depends only on lag, not time
  
- **I(1)**: Series that becomes stationary after **first differencing**
  - $Y_t \sim I(1)$ if $\Delta Y_t = Y_t - Y_{t-1} \sim I(0)$
  - Most common in economics: prices, exchange rates, GDP levels
  
- **I(d)**: General case requiring $d$ differences

**Economic Interpretation:**
- **I(0)**: Returns, growth rates, inflation (fluctuate around a constant mean)
- **I(1)**: Prices, exchange rates, GDP levels (wander without bound)

---

### **11.2 Random Walk Models**

The simplest **I(1)** process is the **random walk**:

**Random Walk (No Drift):**
$$Y_t = Y_{t-1} + \varepsilon_t, \quad \varepsilon_t \sim \text{i.i.d.}(0, \sigma^2)$$

**Properties:**
- $Y_t = Y_0 + \sum_{i=1}^{t} \varepsilon_i$ (cumulative sum of shocks)
- $\mathbb{E}[Y_t] = Y_0$ (constant mean)
- $\text{Var}(Y_t) = t\sigma^2$ ✗ **Non-stationary** (variance grows with time!)
- $\Delta Y_t = \varepsilon_t \sim I(0)$ ✓ **First difference is stationary**

**Random Walk with Drift:**
$$Y_t = \mu + Y_{t-1} + \varepsilon_t$$

**Properties:**
- $Y_t = Y_0 + \mu t + \sum_{i=1}^{t} \varepsilon_i$
- Mean: $\mathbb{E}[Y_t] = Y_0 + \mu t$ ✗ **Non-stationary** (trending mean)
- Variance: $\text{Var}(Y_t) = t\sigma^2$ ✗ **Non-stationary**
- $\Delta Y_t = \mu + \varepsilon_t \sim I(0)$ ✓ **First difference is stationary around drift**

---

### **11.3 Deterministic vs Stochastic Trends**

**Deterministic Trend:**
$$Y_t = \beta_0 + \beta_1 t + \varepsilon_t, \quad \varepsilon_t \sim I(0)$$

- Trend is **fixed** (deterministic function of time)
- Removing trend: $Y_t - \beta_1 t = \beta_0 + \varepsilon_t$ is stationary
- **Trend-stationary**: I(0) after detrending

**Stochastic Trend (Unit Root):**
$$Y_t = Y_{t-1} + \varepsilon_t = Y_0 + \sum_{i=1}^{t} \varepsilon_i$$

- Trend is **random** (cumulative sum of shocks)
- Removing deterministic trend **does not** make it stationary
- **Difference-stationary**: I(1), requires differencing

**Key Distinction:**
- **Deterministic trend**: Detrend → stationary
- **Stochastic trend**: Difference → stationary

**Why It Matters:**
- **Forecasting**: Deterministic trends predictable; stochastic trends not
- **Spurious regression**: Two I(1) series can appear correlated even if independent
- **Cointegration**: Two I(1) series may share a common stochastic trend

---

### **11.4 Visual Intuition: EUR/USD Exchange Rate**

Let's examine the EUR/USD exchange rate to see non-stationarity in action.
