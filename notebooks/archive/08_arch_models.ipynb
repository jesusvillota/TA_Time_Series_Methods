<h2 id="section-1" class="styled-header">1) Introduction to ARCH Models</h2>

**ARCH(q)** (Autoregressive Conditional Heteroskedasticity) models allow volatility to depend on past squared innovations:

$$\sigma_t^2 = \omega + \alpha_1 \varepsilon_{t-1}^2 + \cdots + \alpha_q \varepsilon_{t-q}^2$$

**Key Features**:
- Volatility is **time-varying** and depends on past shocks
- Captures **volatility clustering**
- Requires many lags (q) to capture persistence

**Limitation**: ARCH models require many parameters. **GARCH** models provide a more parsimonious alternative.
