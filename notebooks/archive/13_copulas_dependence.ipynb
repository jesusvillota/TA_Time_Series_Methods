<h2 id="section-1" class="styled-header">1) Introduction to Copulas</h2>

**Copulas** provide a framework for modeling **dependence** between random variables separately from their marginal distributions.

**Key Concepts**:

1. **Sklar's Theorem**: Any multivariate distribution can be decomposed into:
   - Marginal distributions
   - Copula function (captures dependence)

2. **Empirical CDF transformation**: Transform data to uniform marginals using empirical CDF

3. **Tail dependence**: Measures dependence in extreme events (crises)

4. **Asymmetric dependence**: Dependence may differ in upper vs lower tails

**Applications**: Risk management, portfolio optimization, stress testing

<h2 id="section-3" class="styled-header">3) Exercise 27 - Asymmetric and Tail Dependence</h2>

**Slide References**: 27/33, 28/33

We analyze asymmetric and tail dependence using exceedance correlations between equity index returns.

### 3.1 Exceedance Correlations

```stata
* Load all euro area indices
import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ibex35_close
keep date ibex35_close
drop if missing(ibex35_close)
sort date
duplicates drop date, force
tsset date
gen log_ibex = ln(ibex35_close)
gen ribex35 = 100 * D.log_ibex
keep date ribex35
tempfile ibex35
save `ibex35'

* CAC 40
import delimited "$raw_data/CAC40.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close cac40_close
keep date cac40_close
drop if missing(cac40_close)
sort date
duplicates drop date, force
tsset date
gen log_cac = ln(cac40_close)
gen rcac40 = 100 * D.log_cac
keep date rcac40
tempfile cac40
save `cac40'

* DAX
import delimited "$raw_data/DAX.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close dax_close
keep date dax_close
drop if missing(dax_close)
sort date
duplicates drop date, force
tsset date
gen log_dax = ln(dax_close)
gen rdax = 100 * D.log_dax
keep date rdax
tempfile dax
save `dax'

* FTSE MIB
import delimited "$raw_data/FTSEMIB.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ftsemib_close
keep date ftsemib_close
drop if missing(ftsemib_close)
sort date
duplicates drop date, force
tsset date
gen log_ftse = ln(ftsemib_close)
gen rftsemib = 100 * D.log_ftse
keep date rftsemib
tempfile ftsemib
save `ftsemib'

* Merge all indices
use `ibex35', clear
merge 1:1 date using `cac40', nogen keep(match)
merge 1:1 date using `dax', nogen keep(match)
merge 1:1 date using `ftsemib', nogen keep(match)
keep if date >= td(01jan1998) & date <= td(01jan2014)

* Standardize returns (using overall standard deviation)
quietly summarize ribex35
gen std_ribex35 = ribex35 / r(sd)

quietly summarize rcac40
gen std_rcac40 = rcac40 / r(sd)

quietly summarize rdax
gen std_rdax = rdax / r(sd)

quietly summarize rftsemib
gen std_rftsemib = rftsemib / r(sd)

* Calculate exceedance correlations (lower tail)
* Lower tail (both returns < 10th percentile)
_pctile std_ribex35, p(10)
scalar p10_es = r(r1)
_pctile std_rcac40, p(10)
scalar p10_fr = r(r1)
_pctile std_rdax, p(10)
scalar p10_de = r(r1)
_pctile std_rftsemib, p(10)
scalar p10_it = r(r1)

display "Lower Tail Dependence (10th percentile):"
quietly correlate std_ribex35 std_rftsemib if std_ribex35 < p10_es & std_rftsemib < p10_it
display "  ES-IT correlation in lower tail: " r(rho)

quietly correlate std_ribex35 std_rcac40 if std_ribex35 < p10_es & std_rcac40 < p10_fr
display "  ES-FR correlation in lower tail: " r(rho)

quietly correlate std_ribex35 std_rdax if std_ribex35 < p10_es & std_rdax < p10_de
display "  ES-DE correlation in lower tail: " r(rho)
display ""
```

### 3.2 Tail Dependence Analysis

**Interpretation**:

- **Tail dependence**: Measures correlation in extreme events (crises)
- **Lower tail**: Both assets experiencing large negative returns simultaneously
- **Higher tail correlation**: Indicates that during crises, assets move together more than in normal times
- **Asymmetric dependence**: Tail correlations may differ from overall correlation

**Full exceedance correlation analysis** would require calculating correlations at various quantile thresholds and plotting them. This provides a comprehensive view of how dependence varies across the distribution.

<h2 id="section-4" class="styled-header">4) Exercise 28 - Copulas: CDF Transforms</h2>

**Slide Reference**: 31/33

We transform returns to uniform marginals using empirical CDF and visualize copula relationships via scatter plots.

### 4.1 Empirical CDF Transformation

```stata
* Transform to empirical CDF (probability integral transform)
egen rank_es = rank(ribex35)
gen cdf_es = rank_es / (_N + 1)

egen rank_fr = rank(rcac40)
gen cdf_fr = rank_fr / (_N + 1)

egen rank_de = rank(rdax)
gen cdf_de = rank_de / (_N + 1)

egen rank_it = rank(rftsemib)
gen cdf_it = rank_it / (_N + 1)
```

### 4.2 Copula Scatter Plots

```stata
* Scatter plots of CDF transforms
scatter cdf_de cdf_es, ///
    title("Copula: Spain vs Germany") ///
    ytitle("Germany CDF") xtitle("Spain CDF") ///
    name(copula_es_de, replace)

scatter cdf_it cdf_es, ///
    title("Copula: Spain vs Italy") ///
    ytitle("Italy CDF") xtitle("Spain CDF") ///
    name(copula_es_it, replace)

scatter cdf_fr cdf_es, ///
    title("Copula: Spain vs France") ///
    ytitle("France CDF") xtitle("Spain CDF") ///
    name(copula_es_fr, replace)

graph combine copula_es_de copula_es_it copula_es_fr, ///
    title("Copula Scatter Plots") ///
    name(copulas_combined, replace)
```

**Interpretation**:

- **CDF transformation**: Converts data to uniform marginals (values between 0 and 1)
- **Copula scatter plots**: Show dependence structure independent of marginal distributions
- **Patterns**: 
  - **Uniform scatter**: Independence
  - **Concentration along diagonal**: Positive dependence
  - **Concentration in corners**: Tail dependence
- **Asymmetric patterns**: May indicate asymmetric dependence (e.g., stronger in lower tail)

### 5.1 Note on Contour Plots

**Note**: Contour plots of copula densities require specialized commands or custom programming. This analysis is typically performed in R or Python using packages like:

- **R**: `copula` package
- **Python**: `copulae` or visualization libraries

**What Exercise 29 would involve**:

   - Independent normals
   - Correlated normals (Gaussian copula)
   - Symmetric mixture
   - Asymmetric mixture

2. **Generating contour plots**: Visualizing copula densities for each specification

3. **Comparing specifications**: Assessing which copula best captures the dependence structure

**Key Insights**:

- **Gaussian copula**: Assumes symmetric dependence
- **Student t copula**: Captures tail dependence
- **Asymmetric copulas**: Allow for different dependence in upper vs lower tails

**In Stata**: While Stata can perform copula analysis, contour plots are more easily generated in R or Python. For this course, we focus on the empirical CDF transformation and scatter plots (Exercise 28) which provide valuable insights into the dependence structure.
