<h2 class="styled-header">1) Introduction to Cointegration</h2>

### 1.1 What is Cointegration?

**Cointegration** is a fundamental concept in time series econometrics that describes the long-run equilibrium relationship between two or more non-stationary variables. When variables are cointegrated, they share a common stochastic trend and cannot drift arbitrarily far apart over time.

<div style="background: #f0f9ff; border-left: 4px solid #3b82f6; padding: 18px; border-radius: 8px; margin: 20px 0;">

**üîç Formal Definition**

Two time series $X_t$ and $Y_t$ are **cointegrated** of order $(d, b)$, denoted $X_t, Y_t \sim CI(d,b)$, if:

1. Both series are integrated of order $d$: $X_t \sim I(d)$ and $Y_t \sim I(d)$
2. There exists a linear combination that is integrated of order $d-b$ where $b > 0$:
   $$Z_t = Y_t - \beta X_t \sim I(d-b)$$
   
The most common case is $CI(1,1)$ where both series are $I(1)$ but their linear combination is $I(0)$ (stationary).

</div>

**Mathematical Representation:**

If $X_t$ and $Y_t$ are both $I(1)$ processes (unit root processes), they can be written as:

$$
\begin{aligned}
X_t &= X_{t-1} + \varepsilon_{1t} \\
Y_t &= Y_{t-1} + \varepsilon_{2t}
\end{aligned}
$$

If cointegrated with cointegrating vector $(1, -\beta)$, then:

$$
Z_t = Y_t - \beta X_t \sim I(0)
$$

The parameter $\beta$ is called the **cointegrating coefficient**, and $Z_t$ is the **error correction term** or **cointegrating residual**.

### 1.2 The GDP-GDI Relationship

**GDP (Gross Domestic Product)** and **GDI (Gross Domestic Income)** represent two sides of the same national accounting identity. In theory, they should be equal:

$$
\text{GDP} = \text{GDI}
$$

**Why?** 
- **GDP**: Measures total value of goods and services produced (production approach)
- **GDI**: Measures total income generated by production (income approach)
- By definition, every dollar of production generates a dollar of income

**In Practice:**
- Measurement errors create statistical discrepancy: $\text{GDP} \neq \text{GDI}$
- Both series trend upward over time (non-stationary, $I(1)$)
- If cointegrated with $\beta \approx 1$, this confirms the theoretical identity
- Deviations from equilibrium ($Z_t = \text{GDI}_t - \beta \cdot \text{GDP}_t$) should be temporary

<div style="background: linear-gradient(135deg, #fef3c7 0%, #fde68a 100%); border-left: 4px solid #f59e0b; padding: 18px; border-radius: 8px; margin: 20px 0;">

**‚ö†Ô∏è The Spurious Regression Problem**

If we regress one $I(1)$ series on another $I(1)$ series using OLS:

$$Y_t = \alpha + \beta X_t + u_t$$

Without cointegration:
- Standard errors are invalid ‚Üí t-statistics are misleading
- $R^2$ tends to be artificially high
- Residuals $u_t$ are non-stationary ‚Üí violates OLS assumptions
- We find "significant" relationships that are purely spurious

**Solution:** Test for cointegration first! If cointegrated, use VECM instead of standard regression.

</div>

<div style="background: linear-gradient(135deg, #fff5f5 0%, #ffe5e5 100%); border-left: 4px solid #dc2626; padding: 20px; border-radius: 8px; margin: 20px 0;">

**The Spurious Regression Problem**: One of the most important discoveries in econometrics (Granger & Newbold, 1974) is that regressing one non-stationary series on another can produce **highly misleading results** even when the variables are completely independent.

</div>

To visualize this, we construct two **independent random walks** with no causal relationship whatsoever:

$$
\begin{align*}
x_t &= x_{t-1} + \varepsilon_t, \quad x_0 = 0 \\
y_t &= y_{t-1} + \nu_t, \quad y_0 = 0
\end{align*}
$$

where $\varepsilon_t, \nu_t \stackrel{\text{i.i.d.}}{\sim} \mathcal{N}(0,1)$ are **independent** standard normal shocks.

**Key Properties of Random Walks**:
1. **Non-stationary**: $\mathbb{E}[x_t] = 0$ but $\text{Var}(x_t) = t\sigma^2 \to \infty$ as $t \to \infty$
2. **No mean reversion**: Shocks have permanent effects
3. **Trending behavior**: Series wander away from their starting values

#### The Spurious Regression

Despite being generated independently, when we estimate:

$$y_t = \alpha + \beta x_t + u_t$$

we typically observe:
- **Inflated $R^2$** 
- **Significant t-statistics** (often |t| > 2) suggesting "significant" relationship
- **Low Durbin-Watson statistic** (DW ‚âà 0) indicating strongly serially correlated residuals

<div style="background: #f0f9ff; border-left: 4px solid #3b82f6; padding: 16px; border-radius: 6px; margin: 16px 0;">

**Why This Happens**: Both series drift randomly over time. By chance, they may drift in similar directions during certain periods, creating the **illusion** of a relationship. The residuals $\hat{u}_t$ inherit the non-stationarity of $x_t$ and $y_t$, violating the basic assumption that errors are stationary with constant variance.

</div>

```stata
* ---------------------------------------------------------------------------
* Spurious Regression Demonstration: Two Independent Random Walks
* ---------------------------------------------------------------------------

clear all
set obs 250                    // Sample size (250 observations)
set seed 12345                 // Seed that produces more typical spurious result

* Generate time index and declare as time series
gen t = _n
tsset t                        // Declare time series structure

* Generate independent random shocks: Œµ_t, ŒΩ_t ~ N(0,1)
gen eps = rnormal()
gen nu = rnormal()

* Construct random walks: x_t = x_{t-1} + Œµ_t, y_t = y_{t-1} + ŒΩ_t
gen x_rw = sum(eps)            // Cumulative sum creates random walk
gen y_rw = sum(nu)             // Completely independent of x_rw

* Display summary statistics for the random walks
display _newline "{txt}{hline 60}"
display "{txt}Summary Statistics for Independent Random Walks"
display "{txt}{hline 60}"
summarize x_rw y_rw

* Correlation between contemporaneous values (spurious!)
display _newline "{txt}Correlation between independent random walks:"
correlate x_rw y_rw

* Run the spurious regression: y_t = Œ± + Œ≤¬∑x_t + u_t
display _newline "{txt}{hline 60}"
display "{txt}Spurious Regression: y_t = Œ± + Œ≤¬∑x_t + u_t"
display "{txt}{hline 60}"
quietly regress y_rw x_rw

* Calculate Durbin-Watson statistic explicitly
quietly estat dwatson
local dw = r(dw)

* Extract key statistics
local beta = _b[x_rw]
local se_beta = _se[x_rw]
local t_stat = `beta'/`se_beta'
local r2 = e(r2)
local N = e(N)

* Display full regression output
regress y_rw x_rw

* Display interpretation
display _newline "{txt}{hline 60}"
display "{txt}Spurious Regression Results Summary:"
display "{txt}{hline 60}"
display "{txt}Coefficient (Œ≤):     {res}" %7.4f `beta' "{txt}  (should be ~0 if no relation)"
display "{txt}Standard Error:      {res}" %7.4f `se_beta'
display "{txt}t-statistic:         {res}" %7.3f `t_stat' "{txt}  (appears 'significant'!)"
display "{txt}R-squared:           {res}" %7.4f `r2' "{txt}  (" %4.1f `r2'*100 "% of variation 'explained')"
display "{txt}Durbin-Watson:       {res}" %7.4f `dw' "{txt}  (DW‚âà0 indicates strong serial correlation)"
display "{txt}{hline 60}"
display "{err}‚ö†  WARNING: Despite statistical significance,"
display "{err}   these series are INDEPENDENT by construction!"
display "{err}   This demonstrates the SPURIOUS REGRESSION problem."
display "{txt}{hline 60}" _newline

* Generate predictions and residuals for visualization
predict yhat, xb
predict resid, resid

* VISUALIZATION 1: Time series plots of both random walks
twoway (line x_rw t, lcolor(navy) lwidth(medium)) ///
       (line y_rw t, lcolor(maroon) lwidth(medium)), ///
    title("Two Independent Random Walks", size(medium) color(black)) ///
    subtitle("Both series drift randomly with no causal link", size(small)) ///
    ytitle("Value", size(small)) xtitle("Time", size(small)) ///
    legend(order(1 "x{subscript:t} (random walk 1)" 2 "y{subscript:t} (random walk 2)") ///
           size(small) position(6) rows(1)) ///
    name(sr_timeseries, replace)

* VISUALIZATION 2: Scatter plot with spurious regression line
twoway (scatter y_rw x_rw, msymbol(circle) msize(small) mcolor(navy%40)) ///
       (lfit y_rw x_rw, lcolor(red) lwidth(medthick)), ///
    title("Spurious Regression: y{subscript:t} vs x{subscript:t}", size(medium) color(black)) ///
    subtitle("R¬≤ = `: display %4.3f `r2'' with t = `: display %4.2f `t_stat'' despite independence", size(small)) ///
    ytitle("y{subscript:t} (random walk 2)", size(small)) ///
    xtitle("x{subscript:t} (random walk 1)", size(small)) ///
    legend(order(1 "Observed pairs" 2 "OLS fit: Œ≤ = `: display %5.3f `beta''") ///
           size(small) position(6) rows(1)) ///
    note("False relationship due to both series being non-stationary", size(vsmall)) ///
    name(sr_scatter, replace)

* VISUALIZATION 3: Residual plot showing non-stationary behavior
twoway (line resid t, lcolor(red) lwidth(medium)) ///
       (line yhat t, lcolor(blue%30) lwidth(thin) lpattern(dash)), ///
    title("Non-Stationary Residuals from Spurious Regression", size(medium) color(black)) ///
    subtitle("Residuals drift and exhibit strong persistence (DW = `: display %4.3f `dw'')", size(small)) ///
    ytitle("Value", size(small)) xtitle("Time", size(small)) ///
    yline(0, lcolor(gs10) lpattern(solid)) ///
    legend(order(1 "Residuals (√ª{subscript:t})" 2 "Fitted values (≈∑{subscript:t})") ///
           size(small) position(6) rows(1)) ///
    name(sr_resid, replace)

* VISUALIZATION 4: ACF of residuals (should show strong persistence)
ac resid, lags(40) ///
    title("ACF of Residuals: Evidence of Non-Stationarity", size(medium)) ///
    subtitle("Slow decay indicates residuals inherit the unit root", size(small)) ///
    name(sr_acf_resid, replace)
```

### 1.3 Why Cointegration Matters

Cointegration has profound implications for economic modeling and policy:

1. **Long-run Equilibrium**: Identifies stable relationships between variables
2. **Error Correction**: Shows how variables adjust to restore equilibrium
3. **Forecasting**: Improves long-horizon predictions by incorporating equilibrium constraint
4. **Policy Analysis**: Reveals which variables adjust when disequilibrium occurs
5. **Economic Theory**: Tests theoretical relationships (like GDP = GDI)

**Key Distinction:**

| Concept | Description | Example |
|---------|-------------|---------|
| **Correlation** | Short-run co-movement | Two stocks move together daily |
| **Cointegration** | Long-run equilibrium | GDP and GDI cannot diverge permanently |

<h2 class="styled-header">3) Unit Root Testing: Are Both Series I(1)?</h2>

Before testing for cointegration, we must verify that **both series are integrated of the same order**. The Johansen cointegration test requires both variables to be $I(1)$ processes.

<div style="background: #f0f9ff; border-left: 4px solid #3b82f6; padding: 18px; border-radius: 8px; margin: 20px 0;">

**üîç Why Unit Root Tests Matter**

For cointegration to exist between $X_t$ and $Y_t$:
1. Both must be non-stationary (typically $I(1)$)
2. A linear combination must be stationary ($I(0)$)

If both series are already $I(0)$, standard VAR modeling is appropriate‚Äîno need for VECM.

**Key Concept**: An $I(1)$ process becomes stationary after first-differencing: if $X_t \sim I(1)$, then $\Delta X_t \sim I(0)$.

</div>

### 3.1 Testing GDP for Unit Root

We'll use the **DF-GLS test** (Elliott, Rothenberg, and Stock, 1996), which has better power than the standard ADF test, especially for small samples. The test regression is:

$$
\Delta y_t^d = \alpha y_{t-1}^d + \sum_{i=1}^{p} \beta_i \Delta y_{t-i}^d + \varepsilon_t
$$

where $y_t^d$ is the detrended series.

**Hypotheses:**
- $H_0: \alpha = 0$ (unit root exists, series is non-stationary)
- $H_1: \alpha < 0$ (no unit root, series is stationary)

**Decision Rule**: If test statistic < critical value ‚Üí reject $H_0$ ‚Üí stationary

```stata
* DF-GLS unit root test for GDP (levels)
* Test with different lag lengths to check robustness
display _newline(2) "======================================================================"
display "DF-GLS Unit Root Test: GDP (Levels)"
display "======================================================================"

* Lag 1
dfgls gdp, maxlag(1) notrend
display _newline(1) "Lag 1: Test statistic = " %8.4f r(Zt_gls)

* Lag 2
dfgls gdp, maxlag(2) notrend
display _newline(1) "Lag 2: Test statistic = " %8.4f r(Zt_gls)

* Lag 3
dfgls gdp, maxlag(3) notrend
display _newline(1) "Lag 3: Test statistic = " %8.4f r(Zt_gls)

display _newline(1) "Critical Values (MacKinnon):"
display "  1%: -2.58    5%: -1.94    10%: -1.62"
display _newline(1) "Interpretation: Fail to reject H0 ‚Üí GDP has unit root (I(1))"
display "======================================================================"
```

### 3.2 Testing GDI for Unit Root

We apply the same DF-GLS procedure to GDI. If both GDP and GDI fail to reject the unit root null hypothesis, we can proceed with cointegration testing.

```stata
* DF-GLS unit root test for GDI (levels)
display _newline(2) "======================================================================"
display "DF-GLS Unit Root Test: GDI (Levels)"
display "======================================================================"

* Lag 1
dfgls gdi, maxlag(1) notrend
display _newline(1) "Lag 1: Test statistic = " %8.4f r(Zt_gls)

* Lag 2
dfgls gdi, maxlag(2) notrend
display _newline(1) "Lag 2: Test statistic = " %8.4f r(Zt_gls)

* Lag 3
dfgls gdi, maxlag(3) notrend
display _newline(1) "Lag 3: Test statistic = " %8.4f r(Zt_gls)

display _newline(1) "Critical Values (MacKinnon):"
display "  1%: -2.58    5%: -1.94    10%: -1.62"
display _newline(1) "Interpretation: Fail to reject H0 ‚Üí GDI has unit root (I(1))"
display "======================================================================"
```

### 3.3 ACF/PACF Analysis

The **autocorrelation function (ACF)** and **partial autocorrelation function (PACF)** provide visual evidence of non-stationarity:
- **I(1) processes**: ACF decays very slowly, remains high for many lags
- **I(0) processes**: ACF decays quickly to zero

High persistence in ACF confirms unit root test results.

<h2 class="styled-header">4) Testing for Cointegration: Johansen Test</h2>

### 4.1 Johansen Cointegration Test

The **Johansen test** (Johansen, 1988, 1991) is the most widely used multivariate cointegration test. It determines:
1. **Number of cointegrating relationships** (rank of cointegration matrix)
2. **Cointegrating vectors** that define long-run equilibrium

<div style="background: #f0f9ff; border-left: 4px solid #3b82f6; padding: 18px; border-radius: 8px; margin: 20px 0;">

**üîç Johansen Test Framework**

The test is based on a **VAR(p)** model that can be written in error correction form:

$$
\Delta Y_t = \Pi Y_{t-1} + \sum_{i=1}^{p-1} \Gamma_i \Delta Y_{t-i} + \varepsilon_t
$$

where:
- $Y_t := [\text{GDP}_t, \text{GDI}_t]^{\top}$ is a 2√ó1 vector
- $\Pi = \alpha \beta'$ is the long-run impact matrix (2√ó2)
- **$\mathrm{rank}(\Pi$)** determines the number of cointegrating relationships
- $\beta$ contains the cointegrating vectors
- $\alpha$ contains the adjustment coefficients (loading matrix)

**Key Insight**: rank($\Pi$) = 0 ‚Üí no cointegration; rank($\Pi$) = 1 ‚Üí one cointegrating relationship

</div>

**Two Test Statistics:**

1. **Trace Test**: Tests $H_0$: rank ‚â§ $r$ vs $H_1$: rank = 2
   $$\text{Trace} = -T \sum_{i=r+1}^{2} \ln(1 - \hat{\lambda}_i)$$

2. **Maximum Eigenvalue Test**: Tests $H_0$: rank = $r$ vs $H_1$: rank = $r+1$
   $$\text{Max-Eigen} = -T \ln(1 - \hat{\lambda}_{r+1})$$

where $\hat{\lambda}_i$ are estimated eigenvalues (ordered: $\hat{\lambda}_1 > \hat{\lambda}_2$).

### 4.2 Lag Selection for VAR

Before running Johansen test, we must select appropriate lag length for the underlying VAR model using information criteria.

```stata
* Determine optimal lag length for VAR using information criteria
varsoc gdp gdi, maxlag(20)

display _newline(2) "Based on AIC and BIC, we select lag order for subsequent analysis"
```

### 4.2 Interpreting Test Results

The Johansen test produces two test statistics for each rank hypothesis:

**Trace Test Interpretation:**
- Tests $H_0$: rank $\leq r$ against $H_1$: rank $= k$ (full rank)
- Start with $r = 0$: If trace statistic > critical value ‚Üí reject $H_0$ ‚Üí at least one cointegrating relationship
- Continue testing $r = 1, 2, \ldots$ until we fail to reject

**Maximum Eigenvalue Test Interpretation:**
- Tests $H_0$: rank $= r$ against $H_1$: rank $= r+1$
- More powerful for testing specific rank values
- If max-eigen statistic > critical value ‚Üí reject $H_0$ ‚Üí rank is at least $r+1$

<div style="background: #f0f9ff; border-left: 4px solid #3b82f6; padding: 18px; border-radius: 8px; margin: 20px 0;">

**üîç Decision Rule**

1. **Start with rank = 0**: If both tests reject ‚Üí evidence of cointegration
2. **Check rank = 1**: If both tests fail to reject ‚Üí exactly one cointegrating relationship
3. **For our 2-variable system**: Maximum rank = 1 (if both variables are I(1))

**Typical Output:**
- Trace test for rank = 0: Large statistic ‚Üí reject ‚Üí cointegration exists
- Trace test for rank ‚â§ 1: Small statistic ‚Üí fail to reject ‚Üí rank = 1
- Max-eigen test confirms: rank = 1

</div>

```stata
* Johansen cointegration test
* Using lag=2 based on information criteria (adjust if varsoc suggests different lag)
display _newline(2) "======================================================================"
display "JOHANSEN COINTEGRATION TEST: GDP and GDI"
display "======================================================================"

vecrank gdp gdi, lags(3) trend(trend) max

display _newline(2) "Interpretation:"
display "  ‚Ä¢ Trace test: Tests H0: rank ‚â§ r"
display "  ‚Ä¢ Max-Eigen test: Tests H0: rank = r"
display "  ‚Ä¢ If test statistic > 5% critical value ‚Üí reject H0"
display "======================================================================"
```

<div style="background: linear-gradient(135deg, #fef3c7 0%, #fde68a 100%); border-left: 4px solid #f59e0b; padding: 18px; border-radius: 8px; margin: 20px 0;">

**‚ö†Ô∏è Interpreting Johansen Test Results**

After running the Johansen test, examine the output:

1. **Trace test statistics**: Compare to 5% critical values
   - If trace(0) > critical value ‚Üí reject $H_0$: rank = 0 ‚Üí cointegration exists
   - If trace(1) < critical value ‚Üí fail to reject $H_0$: rank ‚â§ 1 ‚Üí rank = 1

2. **Max-eigen test statistics**: Confirm rank determination
   - If max-eigen(0) > critical value ‚Üí at least one cointegrating relationship
   - If max-eigen(1) < critical value ‚Üí exactly one relationship

3. **Cointegrating vector**: Shows the long-run relationship
   - For GDP-GDI, expect $\beta_1 \approx 1$ (one-to-one relationship)

**Conclusion**: If both tests indicate rank = 1, proceed to estimate VECM with one cointegrating relationship.

</div>

<h2 class="styled-header">5) The Cointegrating Relationship</h2>

### 5.1 Estimating the Cointegrating Vector

If the Johansen test confirms cointegration (rank = 1), we can estimate the **cointegrating vector** $\beta$ that defines the long-run equilibrium relationship:

$$
\text{GDI}_t = \beta_0 + \beta_1 \cdot \text{GDP}_t + Z_t
$$

where $Z_t$ is the **error correction term** (cointegrating residual). In equilibrium, $Z_t = 0$; deviations represent temporary disequilibria that should correct over time.

<div style="background: #f0f9ff; border-left: 4px solid #3b82f6; padding: 18px; border-radius: 8px; margin: 20px 0;">

**üîç Economic Interpretation**

For GDP and GDI:
- **Theory predicts**: $\beta_1 \approx 1$ (one-to-one long-run relationship)
- **If $\beta_1 = 1$**: Every euro of GDP generates one euro of income
- **If $Z_t > 0$**: GDI above equilibrium ‚Üí expect downward adjustment
- **If $Z_t < 0$**: GDI below equilibrium ‚Üí expect upward adjustment

The VECM (next section) reveals *which* variable adjusts to restore equilibrium.

</div>

```stata
* Estimate VECM to extract cointegrating vector
* Using rank(1) based on Johansen test results
vec gdp gdi, lags(3) rank(1)

* The cointegrating vector is displayed in the output above
* Beta coefficients show the long-run relationship
```

<div style="background: #f0f9ff; border-left: 4px solid #3b82f6; padding: 18px; border-radius: 8px; margin: 20px 0;">

**üîç Interpretation of results**

- **Cointegrating equation:**
    $$\begin{align*}
    \left[
        \begin{array}{cc}
        1 \\ -1.000953
        \end{array}
    \right]^{\top}
    \left[
        \begin{array}{ll}
        \ln \text{GDP} \\ \ln \text{GDI}
        \end{array}
    \right]
    + 0.0479096 = 0
    \end{align*}$$

    Or equivalently,
    $$
    \begin{align*}
    \ln \text{GDP} = 1.000953 \cdot \ln \text{GDI} - 0.0479096
    \end{align*}
    $$

- **Interpretation of coefficients:**

    - $\beta_1$ (gdp): $1.0$ (normalized for identification)
    - $\beta_2$ (gdi): $-1.000953$ (statistically significant, $p < 0.001$)
    - Constant: $-0.0479096$ 

- **Economic meaning:**

    - The long-run relationship between GDP and GDI is approximately one-to-one ($1.000953 \approx 1.0$), as economic theory predicts: GDP $\approx$ GDI.
    - The intercept term is close to 0, which suggests that there is no evidence of a persistent measurement gap between GDP and GDI in the long run.

</div>


### 5.2 The Error Correction Term

The **error correction term** (ECT) represents deviations from long-run equilibrium. Let's construct and visualize it:

$$
\text{ECT}_t = Z_t = \text{GDI}_t - \hat{\beta}_0 - \hat{\beta}_1 \cdot \text{GDP}_t
$$

A stationary ECT confirms cointegration‚Äîthe deviations are mean-reverting.

```stata
* Predict the error correction term (cointegrating residual)
predict ce_term, ce /// This does the same thing as the line below
/* gen ce_term = gdp - 0.9906542 * gdi - 8.511849 */

* Plot the error correction term
tsline ce_term, ///
    title("Error Correction Term (Cointegrating Residual)", size(medium)) ///
    ytitle("Deviation from Equilibrium") ///
    xtitle("Quarter") ///
    yline(0, lcolor(red) lpattern(dash)) ///
    tlabel(, format(%tqCY)) ///
    note("Mean-reverting pattern indicates cointegration") ///
    scheme(s2color)
/* graph export "../figures/error_correction_term.png", replace width(1200) */
```

```stata
* Test stationarity of the error correction term
* If ECT is stationary (I(0)), this confirms cointegration
dfgls ce_term, maxlag(2) trend

display _newline(1) "If test statistic < -1.94 (5% critical value), ECT is stationary"
display "Stationary ECT confirms the cointegrating relationship"
```

```stata
* ACF of error correction term (should decay quickly if stationary)
ac ce_term, lags(20) ///
    title("Error Correction Term - Autocorrelation Function", size(medium)) ///
    note("Quick decay confirms stationarity") ///
    scheme(s2color)
/* graph export "../figures/ect_acf.png", replace width(1200) */
```
