# Topic 1: Univariate ARMA Models and Unit Root Tests

**CEMFI Advanced Training School - Methods for Time Series**
**Practical Session 1**

## Learning Objectives
1. Identify and estimate ARMA models
2. Perform diagnostic checks
3. Test for unit roots (ADF, PP, KPSS)

## 1. Motivation
ARMA models help forecast returns. Unit root tests determine if differencing is needed.

```stata
* Setup
clear all
set more off
cd "../../data/processed"
```

```stata
* Load S&P 500 data
* use sp500_data.dta, clear
tsset date
* gen sp500_return = D.ln(sp500_close)
summarize sp500_return, detail
```

```stata
* ACF and PACF
ac sp500_return, lags(20)
pac sp500_return, lags(20)
wntestq sp500_return, lags(10)
```

```stata
* Estimate ARMA models
arima sp500_return, ar(1)
estat ic
estimates store ar1

arima sp500_return, ar(1) ma(1)
estat ic
estimates store arma11

estimates stats ar1 arma11
```

```stata
* Diagnostic checks
predict residuals, residuals
wntestq residuals, lags(10)
estat archlm, lags(1/5)
```

```stata
* Unit root tests on EUR/USD
* use eurusd_data.dta, clear
* gen lneurusd = ln(eurusd)
dfuller lneurusd, lags(10) trend
pperron lneurusd, lags(10) trend
* gen d_lneurusd = D.lneurusd
* dfuller d_lneurusd, lags(10)
```
