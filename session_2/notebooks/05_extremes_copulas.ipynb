```stata
* VaR: Normal distribution
summarize returns, detail
scalar var_95 = r(mean) + r(sd) * invnormal(0.05)
```

```stata
* VaR: Historical simulation
summarize returns, detail
scalar var_95_hist = r(p5)
```

```stata
* VaR: GARCH
* arch returns, arch(1) garch(1)
* predict cond_vol, variance
* gen var_95_garch = cond_mean + sqrt(cond_vol) * invnormal(0.05)
```
