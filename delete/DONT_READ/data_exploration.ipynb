```python
import pandas as pd
```

```python
from pathlib import Path

data_path = Path("data/raw")

# list all files in the data/raw directory
files = list(data_path.glob("*.csv"))


import matplotlib.pyplot as plt

def plot_data(df):
    plt.figure(figsize=(10, 5))
    plt.plot(df.iloc[:,0], df.iloc[:,1])
    plt.show()
    
```

```python
for file in files:
    df = pd.read_csv(file)
    plot_data(df)
```
