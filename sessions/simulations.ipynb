# Section 1: Univariate Models

This section implements various univariate time series models including white noise processes, moving average (MA) models, autoregressive (AR) models, and ARMA models.

## 1.1 White Noise - Uniform

Generate $e_1 = \text{uniform}()$ and $e_2 = e_1 \times e_1$

This simulation demonstrates the generation of uniform white noise and its squared transformation.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate uniform white noise
gen t = _n
gen e1 = runiform()
gen e2 = e1 * e1

* Summary statistics
summarize e1 e2

* Time series plots
tsset t
tsline e1, title("Uniform White Noise (e1)") name(e1_plot, replace)
tsline e2, title("Squared Uniform White Noise (e2)") name(e2_plot, replace)

* Correlogram
ac e1, lags(20) title("ACF of Uniform White Noise") name(acf_e1, replace)
ac e2, lags(20) title("ACF of Squared Uniform White Noise") name(acf_e2, replace)
```

## 1.2 White Noise - Gaussian

Generate $u = \text{normal}(0,1)$

This simulation demonstrates Gaussian (normal) white noise, which is fundamental to time series analysis.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate Gaussian white noise
gen t = _n
gen u = rnormal(0, 1)

* Summary statistics
summarize u

* Time series plot
tsset t
tsline u, title("Gaussian White Noise N(0,1)") name(u_plot, replace)

* Histogram
histogram u, normal title("Distribution of Gaussian White Noise") name(hist_u, replace)

* Correlogram
ac u, lags(20) title("ACF of Gaussian White Noise") name(acf_u, replace)
```

## 1.3 MA(1) - Invertible

Simulate $x_t = u_t + \theta u_{t-1}$ with $\theta=0.8$ and $u_t \sim N(0,1)$, then compute $y_t = u_t - \theta u_{t-1}$

An invertible MA(1) process has $|\theta| < 1$. This property ensures the process can be represented as an infinite AR process.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate Gaussian white noise
gen t = _n
gen u = rnormal(0, 1)
tsset t

* MA(1) - Invertible with theta = 0.8
scalar theta = 0.8
gen x = u + theta * L.u
gen y = u - theta * L.u

* Summary statistics
summarize x y

* Time series plots
tsline x, title("MA(1) Invertible: x_t = u_t + 0.8*u_{t-1}") name(ma1_inv_x, replace)
tsline y, title("MA(1) Invertible: y_t = u_t - 0.8*u_{t-1}") name(ma1_inv_y, replace)

* Correlograms
ac x, lags(20) title("ACF of x_t (MA(1) Invertible)") name(acf_ma1_x, replace)
ac y, lags(20) title("ACF of y_t (MA(1) Invertible)") name(acf_ma1_y, replace)
pac x, lags(20) title("PACF of x_t (MA(1) Invertible)") name(pacf_ma1_x, replace)
```

## 1.4 MA(1) - Not Invertible

Simulate $x_t = u_t + (1/\theta)u_{t-1}$ with $\theta=0.8$

A non-invertible MA(1) process has $|\theta| > 1$. In this case, $1/\theta = 1.25 > 1$.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate Gaussian white noise
gen t = _n
gen u = rnormal(0, 1)
tsset t

* MA(1) - Not Invertible with 1/theta = 1.25
scalar theta = 0.8
scalar theta_inv = 1/theta
gen x = u + theta_inv * L.u

* Summary statistics
summarize x

* Time series plot
tsline x, title("MA(1) Not Invertible: x_t = u_t + 1.25*u_{t-1}") name(ma1_notinv, replace)

* Correlograms
ac x, lags(20) title("ACF of x_t (MA(1) Not Invertible)") name(acf_ma1_notinv, replace)
pac x, lags(20) title("PACF of x_t (MA(1) Not Invertible)") name(pacf_ma1_notinv, replace)
```

## 1.5 AR(1) - Stationary

Simulate $x_t = \alpha x_{t-1} + u_t$ with $\alpha=0.8$, $u_t \sim N(0,1)$, and $x_1=100$

A stationary AR(1) process has $|\alpha| < 1$. The process mean-reverts and has constant variance.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate Gaussian white noise
gen t = _n
gen u = rnormal(0, 1)
tsset t

* AR(1) - Stationary with alpha = 0.8
scalar alpha = 0.8
gen x = .
replace x = 100 in 1
replace x = alpha * L.x + u if t > 1

* Summary statistics
summarize x

* Time series plot
tsline x, title("AR(1) Stationary: x_t = 0.8*x_{t-1} + u_t, x_1=100") name(ar1_stat, replace)

* Correlograms
ac x, lags(20) title("ACF of AR(1) Stationary") name(acf_ar1_stat, replace)
pac x, lags(20) title("PACF of AR(1) Stationary") name(pacf_ar1_stat, replace)
```

## 1.6 AR(1) - Not Stationary

Simulate $x_t = \alpha x_{t-1} + u_t$ with $\alpha=1.0001$ and $u_t \sim N(0,1)$

A non-stationary AR(1) process has $|\alpha| \geq 1$. The process is explosive when $\alpha > 1$.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate Gaussian white noise
gen t = _n
gen u = rnormal(0, 1)
tsset t

* AR(1) - Not Stationary with alpha = 1.0001
scalar alpha = 1.0001
gen x = .
replace x = 0 in 1
replace x = alpha * L.x + u if t > 1

* Summary statistics
summarize x

* Time series plot
tsline x, title("AR(1) Not Stationary: x_t = 1.0001*x_{t-1} + u_t") name(ar1_notstat, replace)

* Correlograms
ac x, lags(20) title("ACF of AR(1) Not Stationary") name(acf_ar1_notstat, replace)
pac x, lags(20) title("PACF of AR(1) Not Stationary") name(pacf_ar1_notstat, replace)
```

## 1.7 AR(1) - Random Walk (Wiener Process)

Simulate $x_t = \alpha x_{t-1} + u_t$ with $\alpha=1$ and $u_t \sim N(0,1)$

A random walk is a special case of AR(1) with $\alpha=1$. It is a unit root process and is non-stationary.

## 1.9 Binary Markov Chain

Simulate with $x_0=0$, $P(x_t=0|x_{t-1}=0)=p=0.7$, $P(x_t=1|x_{t-1}=1)=q=0.8$

Using construction: $x_t=(1-x_{t-1})1(u_t>p)+x_{t-1}1(u_t<q)$

This is a two-state Markov chain with persistence.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate uniform random variable
gen t = _n
gen u = runiform()
tsset t

* Markov chain parameters
scalar p = 0.7  // P(x_t=0|x_{t-1}=0)
scalar q = 0.8  // P(x_t=1|x_{t-1}=1)

* Initialize and simulate Markov chain
gen x = .
replace x = 0 in 1
replace x = (1 - L.x) * (u > p) + L.x * (u < q) if t > 1

* Summary statistics
summarize x
tabulate x

* Time series plot
tsline x, title("Binary Markov Chain (p=0.7, q=0.8)") name(markov, replace)

* Correlograms
ac x, lags(20) title("ACF of Binary Markov Chain") name(acf_markov, replace)
pac x, lags(20) title("PACF of Binary Markov Chain") name(pacf_markov, replace)
```

## 1.10 Markov Chain - Unconditional Probability

Calculate and verify $\pi = \frac{1-q}{2-p-q}$ for the binary Markov chain

The unconditional probability of being in state 1 is derived from the ergodic distribution.

```stata
%%stata
clear all
set seed 12345
set obs 10000  // Larger sample for better approximation

* Generate uniform random variable
gen t = _n
gen u = runiform()
tsset t

* Markov chain parameters
scalar p = 0.7
scalar q = 0.8

* Calculate theoretical unconditional probability
scalar pi_theory = (1 - q) / (2 - p - q)
display "Theoretical pi = " pi_theory

* Simulate Markov chain
gen x = .
replace x = 0 in 1
replace x = (1 - L.x) * (u > p) + L.x * (u < q) if t > 1

* Calculate empirical unconditional probability
summarize x
scalar pi_empirical = r(mean)
display "Empirical pi = " pi_empirical
display "Difference = " abs(pi_theory - pi_empirical)
```

## 1.11 Markov Chain - AR(1) Form

Simulate $x_t = 1-p + (p+q-1)x_{t-1} + w_t$ and compare with Markov chain simulation

The Markov chain can be represented in AR(1) form with a specific error structure.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate uniform random variable
gen t = _n
gen u = runiform()
tsset t

* Markov chain parameters
scalar p = 0.7
scalar q = 0.8
scalar c = 1 - p
scalar rho = p + q - 1

* Simulate original Markov chain
gen x_mc = .
replace x_mc = 0 in 1
replace x_mc = (1 - L.x_mc) * (u > p) + L.x_mc * (u < q) if t > 1

* Simulate AR(1) form
gen x_ar = .
replace x_ar = 0 in 1

* Calculate w_t based on AR(1) representation
gen w = x_mc - c - rho * L.x_mc if t > 1
replace x_ar = c + rho * L.x_ar + w if t > 1

* Compare the two representations
summarize x_mc x_ar
correlate x_mc x_ar

* Plot comparison
line x_mc x_ar t in 1/100, title("Markov Chain vs AR(1) Representation") ///
    legend(label(1 "Markov Chain") label(2 "AR(1) Form")) name(compare_mc_ar, replace)
```

# Section 2: Multivariate Models

## 2.1 Multivariate White Noise - Cholesky

Generate $v_t \sim N(0,\Sigma)$ using Cholesky factorization with $\Sigma=\begin{pmatrix}1 & 0.6\\0.6 & 1\end{pmatrix}$ and $H_1=\begin{pmatrix}1&0\\0.6&0.8\end{pmatrix}$

The Cholesky decomposition $\Sigma = H_1 H_1'$ allows us to generate correlated normal variables from independent ones.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate independent standard normal variables
gen t = _n
gen z1 = rnormal(0, 1)
gen z2 = rnormal(0, 1)
tsset t

* Cholesky decomposition: H1 = [1, 0; 0.6, 0.8]
* v = H1 * z
gen v1 = z1
gen v2 = 0.6 * z1 + 0.8 * z2

* Verify covariance structure
correlate v1 v2, covariance

* Time series plots
tsline v1, title("Multivariate White Noise - v1") name(v1_plot, replace)
tsline v2, title("Multivariate White Noise - v2") name(v2_plot, replace)

* Scatter plot
scatter v2 v1, title("Scatter Plot (Cholesky)") name(scatter_chol, replace)

* Correlograms
ac v1, lags(20) title("ACF of v1") name(acf_v1, replace)
ac v2, lags(20) title("ACF of v2") name(acf_v2, replace)
```

## 2.2 Multivariate White Noise - Spectral Decomposition

Generate $v_t$ using spectral decomposition with eigenvalues $\Delta=\text{diag}(1.6,0.4)$

Alternative method using eigenvalue decomposition: $\Sigma = P \Delta P'$

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate independent standard normal variables
gen t = _n
gen z1 = rnormal(0, 1)
gen z2 = rnormal(0, 1)
tsset t

* Eigenvalue decomposition
* Sigma = [[1, 0.6], [0.6, 1]]
* Eigenvalues: 1.6, 0.4
* Eigenvectors (normalized): P = [[0.7071, -0.7071], [0.7071, 0.7071]]
* v = P * sqrt(Delta) * z

scalar sqrt_lambda1 = sqrt(1.6)
scalar sqrt_lambda2 = sqrt(0.4)
scalar p11 = 0.7071
scalar p12 = -0.7071
scalar p21 = 0.7071
scalar p22 = 0.7071

gen v1 = p11 * sqrt_lambda1 * z1 + p12 * sqrt_lambda2 * z2
gen v2 = p21 * sqrt_lambda1 * z1 + p22 * sqrt_lambda2 * z2

* Verify covariance structure
correlate v1 v2, covariance

* Time series plots
tsline v1, title("Multivariate White Noise (Spectral) - v1") name(v1_spec, replace)
tsline v2, title("Multivariate White Noise (Spectral) - v2") name(v2_spec, replace)

* Scatter plot
scatter v2 v1, title("Scatter Plot (Spectral Decomposition)") name(scatter_spec, replace)
```

## 2.3 VMA(1)

Simulate $\begin{pmatrix}x_{1t}\\x_{2t}\end{pmatrix} = \begin{pmatrix}v_{1t}\\v_{2t}\end{pmatrix} + B \begin{pmatrix}v_{1,t-1}\\v_{2,t-1}\end{pmatrix}$ with $B=\begin{pmatrix}0.2 & 0.3\\0.6 & 0.5\end{pmatrix}$

Vector Moving Average of order 1.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate multivariate white noise using Cholesky
gen t = _n
gen z1 = rnormal(0, 1)
gen z2 = rnormal(0, 1)
gen v1 = z1
gen v2 = 0.6 * z1 + 0.8 * z2
tsset t

* VMA(1) with B = [[0.2, 0.3], [0.6, 0.5]]
gen x1 = v1 + 0.2 * L.v1 + 0.3 * L.v2
gen x2 = v2 + 0.6 * L.v1 + 0.5 * L.v2

* Summary statistics
summarize x1 x2
correlate x1 x2, covariance

* Time series plots
tsline x1, title("VMA(1) - x1") name(vma1_x1, replace)
tsline x2, title("VMA(1) - x2") name(vma1_x2, replace)

* Joint plot
scatter x2 x1, title("VMA(1) Scatter Plot") name(scatter_vma1, replace)

* Correlograms
ac x1, lags(20) title("ACF of x1 (VMA(1))") name(acf_vma1_x1, replace)
ac x2, lags(20) title("ACF of x2 (VMA(1))") name(acf_vma1_x2, replace)
```

## 2.4 VAR(1) - Triangular

Simulate $x_t = A x_{t-1} + v_t$ with triangular $A=\begin{pmatrix}0.9 & 0\\1 & 0\end{pmatrix}$

In a triangular VAR system, one variable does not depend on the other contemporaneously.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate multivariate white noise
gen t = _n
gen z1 = rnormal(0, 1)
gen z2 = rnormal(0, 1)
gen v1 = z1
gen v2 = 0.6 * z1 + 0.8 * z2
tsset t

* VAR(1) with triangular A = [[0.9, 0], [1, 0]]
gen x1 = .
gen x2 = .
replace x1 = v1 in 1
replace x2 = v2 in 1
replace x1 = 0.9 * L.x1 + v1 if t > 1
replace x2 = 1.0 * L.x1 + v2 if t > 1

* Summary statistics
summarize x1 x2
correlate x1 x2, covariance

* Time series plots
tsline x1, title("VAR(1) Triangular - x1") name(var1_tri_x1, replace)
tsline x2, title("VAR(1) Triangular - x2") name(var1_tri_x2, replace)

* Joint plot
line x1 x2 t in 1/100, title("VAR(1) Triangular Time Series") name(var1_tri_ts, replace)

* Scatter plot
scatter x2 x1, title("VAR(1) Triangular Scatter") name(scatter_var1_tri, replace)

* Correlograms
ac x1, lags(20) title("ACF of x1 (VAR(1) Triangular)") name(acf_var1_tri_x1, replace)
ac x2, lags(20) title("ACF of x2 (VAR(1) Triangular)") name(acf_var1_tri_x2, replace)
```

## 2.5 VAR(1) - Non-triangular

Simulate $x_t = A x_{t-1} + v_t$ with non-triangular $A=\begin{pmatrix}0.4 & 0.16\\0.36 & 0.4\end{pmatrix}$ using similarity transformation with $P=\begin{pmatrix}0.6 & 0.4\\0 & 1\end{pmatrix}$

A non-triangular VAR has feedback between variables.

```stata
%%stata
clear all
set seed 12345
set obs 1000

* Generate multivariate white noise
gen t = _n
gen z1 = rnormal(0, 1)
gen z2 = rnormal(0, 1)
gen v1 = z1
gen v2 = 0.6 * z1 + 0.8 * z2
tsset t

* VAR(1) with non-triangular A = [[0.4, 0.16], [0.36, 0.4]]
gen x1 = .
gen x2 = .
replace x1 = v1 in 1
replace x2 = v2 in 1
replace x1 = 0.4 * L.x1 + 0.16 * L.x2 + v1 if t > 1
replace x2 = 0.36 * L.x1 + 0.4 * L.x2 + v2 if t > 1

* Summary statistics
summarize x1 x2
correlate x1 x2, covariance

* Time series plots
tsline x1, title("VAR(1) Non-triangular - x1") name(var1_nontri_x1, replace)
tsline x2, title("VAR(1) Non-triangular - x2") name(var1_nontri_x2, replace)

* Joint plot
line x1 x2 t in 1/100, title("VAR(1) Non-triangular Time Series") name(var1_nontri_ts, replace)

* Scatter plot
scatter x2 x1, title("VAR(1) Non-triangular Scatter") name(scatter_var1_nontri, replace)

* Correlograms
ac x1, lags(20) title("ACF of x1 (VAR(1) Non-triangular)") name(acf_var1_nontri_x1, replace)
ac x2, lags(20) title("ACF of x2 (VAR(1) Non-triangular)") name(acf_var1_nontri_x2, replace)
```

## 3.1 Local Level Model with Kalman Filter

Implement Kalman filter for:
- Observation equation: $y_t = \mu_t + u_t$ where $u_t \sim N(0, \sigma_u^2)$
- State equation: $\mu_t = \mu_{t-1} + v_t$ where $v_t \sim N(0, \sigma_v^2)$

With the prediction update formula:
$$y_t^* = y_{t-1}^* + (1-\delta)(y_{t-1}-y_{t-1}^*)$$

where $\delta=\frac{\sigma_u^2}{\sigma_u^2+\sigma_v^2+p}$ and $p = -\sigma_v^2 + \sqrt{\sigma_v^4 + 4\sigma_u^2\sigma_v^2} / 2$

```stata
%%stata
clear all
set seed 12345
set obs 500

* Generate time index
gen t = _n
tsset t

* Set variance parameters
scalar sigma_u2 = 1.0  // Observation noise variance
scalar sigma_v2 = 0.5  // State noise variance

* Generate true unobserved level (random walk)
gen v = rnormal(0, sqrt(sigma_v2))
gen mu_true = .
replace mu_true = 0 in 1
replace mu_true = L.mu_true + v if t > 1

* Generate observations
gen u = rnormal(0, sqrt(sigma_u2))
gen y = mu_true + u

* Calculate Kalman filter parameter delta
scalar p = (-sigma_v2 + sqrt(sigma_v2^2 + 4*sigma_u2*sigma_v2)) / 2
scalar delta = sigma_u2 / (sigma_u2 + sigma_v2 + p)
display "Delta = " delta
display "p = " p

* Initialize Kalman filter
gen y_pred = .
gen mu_filter = .
replace y_pred = y in 1
replace mu_filter = y in 1

* Kalman filter recursion
forvalues i = 2/500 {
    * Prediction step
    quietly replace y_pred = mu_filter[`i'-1] in `i'
    
    * Update step
    quietly replace mu_filter = y_pred[`i'] + (1 - delta) * (y[`i'-1] - y_pred[`i'-1]) in `i'
}

* Plot results
tsline y mu_true mu_filter, ///
    title("Local Level Model with Kalman Filter") ///
    legend(label(1 "Observed y_t") label(2 "True μ_t") label(3 "Filtered μ_t")) ///
    name(kalman_filter, replace)

* Plot first 100 observations for clarity
line y mu_true mu_filter t in 1/100, ///
    title("Kalman Filter - First 100 Observations") ///
    legend(label(1 "Observed") label(2 "True State") label(3 "Filtered")) ///
    name(kalman_zoom, replace)

* Summary statistics
gen filter_error = mu_true - mu_filter
summarize filter_error
display "Mean Squared Error: " r(Var) + r(mean)^2
```
