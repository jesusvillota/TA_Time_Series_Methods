**Data Sources:**
- **FRED (Federal Reserve Economic Data)**: Macroeconomic data, bond yields, VIX, unemployment rates

- **Raw CSV files**: Stored in `/data/raw/`

**Processing Steps:**
1. Load raw CSV files with proper date parsing
2. Convert dates to Stata time series formats (daily, monthly, quarterly)
3. Create derived variables (returns, spreads, growth rates)
4. Handle missing values and duplicates
5. Set up time series structure
6. Add variable labels for clarity

**Note**: This notebook must be run before any other Session 1 notebooks, as it prepares all the datasets used in the exercises.

<h2 class="styled-header">2) Stock Market Indices</h2>

We load four major European stock market indices from Yahoo Finance:
- **IBEX-35**: Spanish stock market index
- **CAC 40**: French stock market index  
- **DAX**: German stock market index
- **FTSE MIB**: Italian stock market index

**Data Format**: Yahoo Finance CSV files have a 3-row header structure, so we use `rowrange(4)` to skip headers and start reading from row 4.

**Transformations**:
- Calculate daily log returns: $r_t = 100 \times (\ln(P_t) - \ln(P_{t-1}))$
- For IBEX-35, also create monthly returns for some exercises
- Create squared returns for volatility analysis


### 2.1 IBEX-35 (Spain)


```stata
* ----------------------------------------------------------------------------
* Load Spanish Stock Index (IBEX-35) - Yahoo Finance Format
* ----------------------------------------------------------------------------

display "Loading IBEX-35 data..."

import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
// clear: Clears any existing data from memory before importing
// varnames(1): Uses row 1 (after skipping rows) as variable names
// rowrange(4): Starts reading from row 4 (skips Yahoo Finance's 3-row header)
// stringcols(_all): Imports all columns as strings for safe parsing

* Parse dates from Yahoo Finance format (row 4 onwards)
gen date_numeric = date(price, "YMD") // Converts date string to Stata numeric date (Year-Month-Day format)
format date_numeric %td // Applies daily date format (%td) for proper display
rename date_numeric date // Renames to "date" for clarity

* Rename and keep relevant variables
destring close, replace force // Converts "close" from string to numeric, forcing conversion
rename close ibex35_close // Renames closing price variable to ibex35_close
keep date ibex35_close // Keeps only date and closing price columns

* Handle missing values
destring ibex35_close, replace force // Ensures numeric format (redundant but safe)
drop if missing(ibex35_close) // Removes rows with missing closing prices

* Sort and check for duplicates
sort date // Sorts dataset by date in ascending order
duplicates drop date, force // Removes duplicate dates, keeping first occurrence

* Set time series
tsset date // Declares dataset as time series with date as time variable

* Create returns (log differences)
gen log_ibex = ln(ibex35_close) // Creates natural log of closing prices
gen ribex35 = 100 * D.log_ibex // Calculates daily log returns as percentage (D. = first difference operator)
drop log_ibex // Removes temporary log price variable
label var ribex35 "Ibex-35 Daily Returns (%)" // Adds descriptive label

* Create squared returns for volatility analysis
gen ribex35sq = ribex35^2 // Squares returns for volatility modeling
label var ribex35sq "Ibex-35 Squared Returns" // Adds descriptive label

* Create monthly returns for some exercises
gen year = year(date) // Extracts year component from daily date
gen month = month(date) // Extracts month component from daily date
gen date_m = ym(year, month) // Creates monthly date variable using ym() function
format date_m %tm // Formats as monthly time variable (%tm)

* Keep last observation of each month for monthly series
preserve // Temporarily saves current dataset state (creates a copy)
bysort date_m: egen temp_last = max(date) // For each month, finds the last (maximum) trading day
keep if date == temp_last // Keeps only observations that are the last day of each month
keep date_m ibex35_close // Keeps only monthly date and closing price
rename ibex35_close ibex35_close_m // Renames to indicate monthly data
tsset date_m // Sets time series structure with monthly frequency
gen log_ibex_m = ln(ibex35_close_m) // Creates natural log of monthly closing prices
gen ribex35_monthly = 100 * D.log_ibex_m // Calculates monthly log returns as percentage
drop log_ibex_m // Removes temporary log price variable
label var ribex35_monthly "Ibex-35 Monthly Returns (%)" // Adds descriptive label
tempfile ibex35_monthly // Creates temporary file handle
save `ibex35_monthly' // Saves monthly data to temporary file
restore // Restores original daily dataset from before preserve

* Save daily IBEX data
tempfile ibex35_daily // Creates temporary file handle for daily data
save `ibex35_daily' // Saves daily data to temporary file for later use

display "  IBEX-35 data loaded successfully"
display "  Date range: " %td r(tmin) " to " %td r(tmax) // Displays min and max dates
display "  Observations: " _N // Displays total number of observations (_N = total rows)
display ""

* ----------------------------------------------------------------------------
* Visualize IBEX-35 Data
* ----------------------------------------------------------------------------

* View the dataframe (first/last observations and summary statistics)
list date ibex35_close ribex35 in 1/10 // Lists first 10 observations
list date ibex35_close ribex35 in -10/L // Lists last 10 observations (L = last row)
summarize ibex35_close ribex35 ribex35sq // Displays summary statistics (mean, sd, min, max)

* Plot price level over time
tsline ibex35_close, /// // tsline: Time series line plot command for date-indexed data
    title("IBEX-35 Price Level") /// // Sets the main title of the graph
    ytitle("Closing Price") xtitle("Date") /// // Sets y-axis and x-axis labels
    name(ibex_price, replace) // Saves graph with name "ibex_price", replacing if exists

* Plot daily returns
tsline ribex35, /// // Creates time series line plot of daily returns
    title("IBEX-35 Daily Returns") /// // Sets the main title
    ytitle("Returns (%)") xtitle("Date") /// // Sets y-axis and x-axis labels
    yline(0, lcolor(red) lpattern(dash)) /// // Adds horizontal reference line at y=0 (red, dashed)
    name(ibex_returns, replace) // Saves graph with name "ibex_returns", replacing if exists

* Plot both together
graph combine ibex_price ibex_returns, /// // Combines two previously created graphs into one figure
    title("IBEX-35: Price and Returns") /// // Sets overall title for combined graph
    rows(2) name(ibex_combined, replace) // Arranges graphs in 2 rows (vertically stacked), saves as "ibex_combined"

* Histogram of returns
histogram ribex35, /// // Creates histogram of return distribution
    title("Distribution of IBEX-35 Returns") /// // Sets the main title
    xtitle("Returns (%)") /// // Sets x-axis label
    normal name(ibex_hist, replace) // Overlays normal distribution curve, saves as "ibex_hist"
```

<h2 class="styled-header">3) Government Bond Yields and Spreads</h2>

We load 10-year government bond yields for four Eurozone countries from FRED:
- **Germany**: Benchmark (used to calculate spreads)
- **Spain**: Used in sovereign spread analysis
- **France**: Used in correlation analysis
- **Italy**: Used in correlation analysis

**Data Format**: FRED CSV files with `observation_date` and value columns.

**Transformations**:
- Convert to monthly frequency (take first observation of each month)
- Calculate sovereign spreads relative to Germany: $\text{spread} = (\text{yield} - \text{yield}_{DE}) \times 100$ (in basis points)
- Create bond returns (first differences) for correlation analysis


```stata
* ----------------------------------------------------------------------------
* Load Government Bond Yields (FRED Format - Monthly)
* ----------------------------------------------------------------------------

display "Loading government bond yields..." // Prints loading message

* Germany 10-Year (Benchmark)
import delimited "$raw_data/BOND_10Y_DE.csv", clear // Imports German bond yield CSV, clears memory
gen date_numeric = date(observation_date, "YMD") // Converts observation_date string to Stata numeric date
format date_numeric %td // Applies daily date format
gen year = year(date_numeric) // Extracts year component from date
gen month = month(date_numeric) // Extracts month component from date
gen date_m = ym(year, month) // Creates monthly date variable using ym() function
format date_m %tm // Formats as monthly time variable (%tm)
drop observation_date date_numeric year month // Drops temporary variables
rename date_m date // Renames to "date" for clarity
rename irltlt01dem156n bond_de // Renames FRED series code to bond_de
destring bond_de, replace force // Converts to numeric, forcing conversion
sort date // Sorts by date
duplicates drop date, force // Removes duplicate dates
tsset date // Sets time series structure with date as time variable
tempfile bond_de // Creates temporary file handle
save `bond_de' // Saves to temporary file
display "  Germany 10Y bond loaded" // Prints confirmation

* Spain 10-Year
import delimited "$raw_data/BOND_10Y_ES.csv", clear // Imports Spanish bond yield CSV, clears memory
gen date_numeric = date(observation_date, "YMD") // Converts observation_date string to Stata numeric date
format date_numeric %td // Applies daily date format
gen year = year(date_numeric) // Extracts year component
gen month = month(date_numeric) // Extracts month component
gen date_m = ym(year, month) // Creates monthly date variable
format date_m %tm // Formats as monthly time variable
drop observation_date date_numeric year month // Drops temporary variables
rename date_m date // Renames to "date"
rename irltlt01esm156n bond_es // Renames FRED series code to bond_es
destring bond_es, replace force // Converts to numeric
sort date // Sorts by date
duplicates drop date, force // Removes duplicates
tsset date // Sets time series
tempfile bond_es // Creates temporary file handle
save `bond_es' // Saves to temporary file
display "  Spain 10Y bond loaded" // Prints confirmation

* France 10-Year
import delimited "$raw_data/BOND_10Y_FR.csv", clear // Imports French bond yield CSV, clears memory
gen date_numeric = date(observation_date, "YMD") // Converts observation_date string to Stata numeric date
format date_numeric %td // Applies daily date format
gen year = year(date_numeric) // Extracts year component
gen month = month(date_numeric) // Extracts month component
gen date_m = ym(year, month) // Creates monthly date variable
format date_m %tm // Formats as monthly time variable
drop observation_date date_numeric year month // Drops temporary variables
rename date_m date // Renames to "date"
rename irltlt01frm156n bond_fr // Renames FRED series code to bond_fr
destring bond_fr, replace force // Converts to numeric
sort date // Sorts by date
duplicates drop date, force // Removes duplicates
tsset date // Sets time series
tempfile bond_fr // Creates temporary file handle
save `bond_fr' // Saves to temporary file
display "  France 10Y bond loaded" // Prints confirmation

* Italy 10-Year
import delimited "$raw_data/BOND_10Y_IT.csv", clear // Imports Italian bond yield CSV, clears memory
gen date_numeric = date(observation_date, "YMD") // Converts observation_date string to Stata numeric date
format date_numeric %td // Applies daily date format
gen year = year(date_numeric) // Extracts year component
gen month = month(date_numeric) // Extracts month component
gen date_m = ym(year, month) // Creates monthly date variable
format date_m %tm // Formats as monthly time variable
drop observation_date date_numeric year month // Drops temporary variables
rename date_m date // Renames to "date"
rename irltlt01itm156n bond_it // Renames FRED series code to bond_it
destring bond_it, replace force // Converts to numeric
sort date // Sorts by date
duplicates drop date, force // Removes duplicates
tsset date // Sets time series
tempfile bond_it // Creates temporary file handle
save `bond_it' // Saves to temporary file
display "  Italy 10Y bond loaded" // Prints confirmation

* Merge all bond yields
use `bond_de', clear // Loads German bond data as base
merge 1:1 date using `bond_es', nogen // Merges Spanish data (1:1 on date, no generate new variable)
merge 1:1 date using `bond_fr', nogen // Merges French data
merge 1:1 date using `bond_it', nogen // Merges Italian data

* Create spreads relative to Germany (in basis points)
gen spread_es = (bond_es - bond_de) * 100 // Calculates Spain-Germany spread in basis points
gen spread_fr = (bond_fr - bond_de) * 100 // Calculates France-Germany spread in basis points
gen spread_it = (bond_it - bond_de) * 100 // Calculates Italy-Germany spread in basis points

label var spread_es "Spain Sovereign Spread (bp)" // Adds label for Spain spread
label var spread_fr "France Sovereign Spread (bp)" // Adds label for France spread
label var spread_it "Italy Sovereign Spread (bp)" // Adds label for Italy spread

* Create bond returns for correlation analysis
gen rbond_es = D.bond_es // Creates first difference (return) for Spain bond yield
gen rbond_fr = D.bond_fr // Creates first difference for France
gen rbond_de = D.bond_de // Creates first difference for Germany
gen rbond_it = D.bond_it // Creates first difference for Italy

* Set time series
tsset date // Declares dataset as time series with date as time variable

* Save bond data
tempfile bonds_monthly // Creates temporary file handle for merged bond data
save `bonds_monthly' // Saves merged data to temporary file

display "  Bond yields merged and spreads calculated" // Prints confirmation
display "" // Prints empty line for spacing
```

```stata
* ----------------------------------------------------------------------------
* Load Spanish GDP (Quarterly - will be converted to growth rate)
* ----------------------------------------------------------------------------

display "Loading Spanish GDP data..." // Prints loading message

import delimited "$raw_data/ESP_GDP.csv", clear // Imports Spanish GDP CSV, clears memory
gen date_numeric = date(observation_date, "YMD") // Converts observation_date string to Stata numeric date
format date_numeric %td // Applies daily date format
gen year = year(date_numeric) // Extracts year component
gen quarter = quarter(date_numeric) // Extracts quarter component
gen date_q = yq(year, quarter) // Creates quarterly date variable using yq() function
format date_q %tq // Formats as quarterly time variable (%tq)
drop observation_date date_numeric year quarter // Drops temporary variables
rename date_q date // Renames to "date" for clarity
rename cpmnacscab1gqes gdp_esp // Renames FRED series code to gdp_esp
destring gdp_esp, replace force // Converts to numeric, forcing conversion
sort date // Sorts by date
duplicates drop date, force // Removes duplicate dates
tsset date // Sets time series structure

* Create annual growth rate of GDP
gen log_gdp = ln(gdp_esp) // Creates natural log of GDP for growth rate calculation
gen dgdp = 100 * (log_gdp - L4.log_gdp) // Calculates year-over-year growth rate (L4 = 4 quarters lag)
label var dgdp "Spain Real GDP Growth Rate (YoY %)" // Adds descriptive label
label var gdp_esp "Spain Real GDP (Level)" // Adds descriptive label

tempfile esp_gdp // Creates temporary file handle
save `esp_gdp' // Saves data to temporary file
display "  Spain GDP loaded" // Prints confirmation
```

### 5.1 VIX Index

The VIX (CBOE Volatility Index) measures market expectations of near-term volatility conveyed by S&P 500 index option prices.

**Data Source**: FRED (VIXCLS series)

**Transformations**:
- Create log-VIX: $\text{lgvix}_t = \ln(VIX_t)$ for modeling purposes (log transformation stabilizes variance)

```stata
* ----------------------------------------------------------------------------
* Load VIX Index (FRED Format)
* ----------------------------------------------------------------------------

display "Loading VIX data..." // Prints loading message

import delimited "$raw_data/VIXCLS.csv", clear // Imports VIX CSV, clears memory

* Parse dates from FRED format
gen date_numeric = date(observation_date, "YMD") // Converts observation_date string to Stata numeric date
format date_numeric %td // Applies daily date format
drop observation_date // Drops original date string
rename date_numeric date // Renames to "date" for clarity
rename vixcls vix // Renames FRED series code to vix

* Handle missing values
destring vix, replace force // Converts to numeric, forcing conversion
drop if missing(vix) // Removes rows with missing VIX values

* Sort and set time series
sort date // Sorts by date
duplicates drop date, force // Removes duplicate dates
tsset date // Sets time series structure

* Create log-VIX for modeling
gen lgvix = ln(vix) // Creates natural log of VIX for variance stabilization
label var lgvix "Log VIX" // Adds descriptive label

* Save VIX data
tempfile vix_daily // Creates temporary file handle
save `vix_daily' // Saves data to temporary file

display "  VIX data loaded successfully" // Prints confirmation
display "  Observations: " _N // Displays number of observations (_N = total rows)
display "" // Prints empty line for spacing
```
