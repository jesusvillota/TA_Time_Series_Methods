<h2 class="styled-header">1) Introduction to Cointegration</h2>

**Cointegration** occurs when two or more non-stationary series share a **common stochastic trend**, such that a linear combination of them is stationary.

**Key Concept**: Even if individual series are non-stationary (have unit roots), their **linear combination** may be stationary if they are cointegrated.

**Example**: Two stock prices may both be non-stationary, but their **spread** (difference) may be stationary if they move together.

**Testing for Cointegration**:
- **Engle-Granger test**: For two variables
- **Johansen test**: For multiple variables (more powerful)

**Vector Error-Correction Model (VECM)**: If variables are cointegrated, we can estimate a VECM that captures both short-run dynamics and long-run equilibrium relationships.

### 2.2 Time Series Plots


```python
* ----------------------------------------------------------------------------
* EXERCISE 9: Cointegration - Unemployment rates example
* Slides 65/69-69/69
* ----------------------------------------------------------------------------

* Load US unemployment data from processed file
use "$processed_data/us_unemployment_monthly.dta", clear

* Convert to quarterly for consistency with slides
gen year = year(dofm(date))
gen quarter = quarter(dofm(date))
gen date_q = yq(year, quarter)
format date_q %tq

* Aggregate to quarterly (take average)
collapse (mean) unemp_ca unemp_tx, by(date_q)
rename date_q date
tsset date

keep if date >= yq(1976, 3) & date <= yq(2012, 4)

* Time series plots
tsline unemp_ca, ///
    title("California Unemployment Rate") ///
    ytitle("Unemployment (%)") xtitle("Date") ///
    name(unemp_ca_ts, replace)

tsline unemp_tx, ///
    title("Texas Unemployment Rate") ///
    ytitle("Unemployment (%)") xtitle("Date") ///
    name(unemp_tx_ts, replace)

graph combine unemp_ca_ts unemp_tx_ts, ///
    title("Regional Unemployment Rates") ///
    name(unemp_combined, replace)
```

### 2.3 Unit Root Tests


```python
* Unit root tests
preserve
drop if missing(unemp_ca)
gen t = _n
tsset t
display "DF-GLS Test for California Unemployment:"
quietly dfgls unemp_ca, maxlag(3) notrend
display "  Test statistic: " r(dfgls1)
display "  Observations: " r(N)
restore

display ""

preserve
drop if missing(unemp_tx)
gen t = _n
tsset t
display "DF-GLS Test for Texas Unemployment:"
quietly dfgls unemp_tx, maxlag(3) notrend
display "  Test statistic: " r(dfgls1)
display "  Observations: " r(N)
restore
```

**Expected Result**: Both series should fail to reject the unit root null (non-stationary), which is a prerequisite for cointegration analysis.


### 2.4 Johansen Cointegration Test


```python
* Johansen cointegration test
display "Johansen Cointegration Test (California and Texas):"
vecrank unemp_ca unemp_tx, lags(3) trend(rconstant)
```

**Interpretation**:
- **Trace statistic**: Tests the null of no cointegration vs. alternative of cointegration
- **Maximum eigenvalue statistic**: Tests the null of $r$ cointegrating relationships vs. $r+1$
- If we **reject** the null, the variables are cointegrated
- The number of cointegrating relationships determines the rank of the VECM


### 2.5 Vector Error-Correction Model (VECM)


```python
* Estimate VECM if cointegrated
display "Vector Error-Correction Model (VECM):"
vec unemp_ca unemp_tx, lags(3) trend(rconstant) rank(1)
```

**VECM Interpretation**:
- **Error-correction term**: Captures deviations from long-run equilibrium
- **Coefficient**: Speed of adjustment back to equilibrium
- **Lagged differences**: Short-run dynamics

**Economic Implication**: If unemployment rates are cointegrated, there is a **long-run equilibrium relationship** between them, even though they may diverge in the short run.
