<h2 class="styled-header">1) Introduction to Autocorrelation</h2>

The **autocorrelation function (ACF)** measures the linear dependence between observations at different time lags. For a time series $\{Y_t\}$, the autocorrelation at lag $h$ is:

$$\rho(h) = \frac{\mathbb{C}\text{ov}(Y_t, Y_{t-h})}{\sqrt{\mathbb{V}(Y_t)\mathbb{V}(Y_{t-h})}} = \frac{\gamma(h)}{\gamma(0)}$$

**Key Properties**:
- $\rho(0) = 1$ (perfect correlation with itself)
- $|\rho(h)| \leq 1$ for all $h$
- For stationary processes, $\rho(h) = \rho(-h)$ (symmetry)

**Sample ACF**: 
$$\hat{\rho}(h) = \frac{\sum_{t=h+1}^{T} (y_t - \bar{y})(y_{t-h} - \bar{y})}{\sum_{t=1}^{T} (y_t - \bar{y})^2}$$

**Statistical Inference**: Under the null hypothesis $H_0: \rho(h) = 0$ for a white noise process:
$$\hat{\rho}(h) \sim \mathcal{N}\left(0, \frac{1}{T}\right)$$

This gives us **95% confidence bands**: $\pm 1.96/\sqrt{T}$ (Bartlett's bands).

**Economic Interpretation**:
- **High autocorrelation**: Past values strongly predict current values (persistence)
- **Low autocorrelation**: Series behaves more like white noise (weak predictability)
- **Negative autocorrelation**: Mean reversion (high values tend to be followed by low values)

In this notebook, we analyze autocorrelation patterns in:
1. **Stock returns** (IBEX-35): Typically show low autocorrelation (weak-form market efficiency)
2. **Sovereign spreads**: May show higher autocorrelation (persistent risk premia)

### 2.1 IBEX-35 Monthly Returns


```stata
* ----------------------------------------------------------------------------
* EXERCISE 1: Autocorrelation - Examples based on Spanish data
* ----------------------------------------------------------------------------

* Load IBEX-35 monthly data from processed data directory
use "$processed_data/ibex35_monthly.dta", clear 
tsset date_m // Ensure time series is set correctly

list date ibex35_close ribex35 in 1/10

* Time series plot
tsline ribex35_monthly, /// Create time series line plot of monthly returns
    title("IBEX-35 Monthly Returns") /// Set plot title
    ytitle("Returns (%)") xtitle("Date") /// Set y-axis label to "Returns (%)" and x-axis label to "Date"
    name(ibex_returns_ts, replace) // Save graph with name "ibex_returns_ts", replace if it already exists

* Autocorrelation
ac ribex35_monthly, lags(40) /// Calculate and plot autocorrelation function for returns with 40 lags (includes Bartlett's confidence bands)
    title("Autocorrelation: IBEX-35 Returns") /// Set plot title
    name(acf_ibex, replace) // Save autocorrelation graph with name "acf_ibex", replace if it already exists

display "IBEX-35 Returns Summary Statistics:" // Display text message in output
summarize ribex35_monthly, detail // Calculate and display detailed summary statistics (mean, std dev, min, max, percentiles) for returns
```

**Bartlett's Confidence Bands**: For a white noise process, the 95% confidence interval for the sample ACF at lag $h$ is approximately:

$$\hat{\rho}(h) \pm 1.96 \times \frac{1}{\sqrt{T}}$$

If autocorrelation coefficients fall outside these bands, we can reject the null hypothesis that $\rho(h) = 0$ at the 5% significance level.

**Interpretation**: The ACF plot shows autocorrelation coefficients at lags 1-40. For stock returns, we typically expect:
- **Low autocorrelation** at all lags (consistent with weak-form market efficiency)
- The **gray bands** represent Bartlett's 95% confidence intervals
- Most autocorrelation coefficients fall **within the confidence bands**, indicating they are **not statistically significant**
- This suggests that past returns have little predictive power for future returns
- Any coefficients outside the bands may be due to **sampling variation** rather than true autocorrelation

**Statistical Test**: If we observe many coefficients outside the confidence bands, we would reject the null hypothesis that the series is white noise. For IBEX-35 returns, we typically find few significant autocorrelations, supporting market efficiency.

**Economic Implication**: If stock returns were predictable from past returns, investors could exploit this to earn excess returns, which would violate the efficient market hypothesis (at least in its weak form).

1. **1993-1999**: Pre-euro convergence period
2. **2000-2007**: Euro adoption and pre-crisis period
3. **2008-2015**: Financial crisis and European debt crisis
4. **2015-2019**: Recovery period
5. **2020-2023**: COVID-19 pandemic period

**Research Question**: Do correlations increase during crisis periods? This would indicate **contagion** or **common risk factors** affecting all countries simultaneously.

**Interpretation of Correlation Patterns**:

- **1993-1999**: Moderate correlations (around 0.75-0.99)
  - Countries were converging toward euro adoption
  - Some divergence in fiscal policies
  
- **2000-2007**: Similar correlations (around 0.73-0.88)
  - Euro adoption reduced currency risk
  - Common monetary policy but different fiscal policies
  
- **2008-2015**: **Higher correlations** (around 0.83-0.95)
  - **Contagion effect**: Crisis in one country affects others
  - **Common risk factors**: European-wide concerns about fiscal sustainability
  - **Flight to quality**: All non-German bonds perceived as riskier relative to Germany

- **2015-2019**: Recovery period correlations
  - Post-crisis normalization as European economies recovered
  - ECB's quantitative easing program reduced spreads and correlations
  - Gradual return to more country-specific risk factors

- **2020-2023**: COVID-19 pandemic period
  - **High correlations expected** due to global shock
  - Synchronized fiscal and monetary policy responses across Europe
  - Common health and economic crisis affecting all countries simultaneously

- **2024-2025**: Post-COVID period
  - Transition period as economies normalize post-pandemic
  - Assessment of whether correlations return to pre-pandemic levels
  - Impact of new European fiscal rules and monetary policy normalization

**Economic Implication**: The increase in correlations during crises suggests that **diversification benefits are reduced** during periods of financial stress, which is exactly when diversification is most needed.
