<h2 id="section-1" class="styled-header">1) Introduction to Autocorrelation</h2>

The **autocorrelation function (ACF)** measures the linear dependence between observations at different time lags. For a time series $\{Y_t\}$, the autocorrelation at lag $h$ is:

$$\rho(h) = \frac{\text{Cov}(Y_t, Y_{t-h})}{\sqrt{\text{Var}(Y_t)\text{Var}(Y_{t-h})}} = \frac{\gamma(h)}{\gamma(0)}$$

**Key Properties**:
- $\rho(0) = 1$ (perfect correlation with itself)
- $|\rho(h)| \leq 1$ for all $h$
- For stationary processes, $\rho(h) = \rho(-h)$ (symmetry)

**Sample ACF**: 
$$\hat{\rho}(h) = \frac{\sum_{t=h+1}^{T} (y_t - \bar{y})(y_{t-h} - \bar{y})}{\sum_{t=1}^{T} (y_t - \bar{y})^2}$$

**Statistical Inference**: Under the null hypothesis $H_0: \rho(h) = 0$ for a white noise process:
$$\hat{\rho}(h) \sim \mathcal{N}\left(0, \frac{1}{T}\right)$$

This gives us **95% confidence bands**: $\pm 1.96/\sqrt{T}$ (Bartlett's bands).

**Economic Interpretation**:
- **High autocorrelation**: Past values strongly predict current values (persistence)
- **Low autocorrelation**: Series behaves more like white noise (weak predictability)
- **Negative autocorrelation**: Mean reversion (high values tend to be followed by low values)

In this notebook, we analyze autocorrelation patterns in:
1. **Stock returns** (IBEX-35): Typically show low autocorrelation (weak-form market efficiency)
2. **Sovereign spreads**: May show higher autocorrelation (persistent risk premia)

### 2.1 IBEX-35 Monthly Returns


```python
* ----------------------------------------------------------------------------
* EXERCISE 1: Autocorrelation - Examples based on Spanish data
* Slide 10/69
* ----------------------------------------------------------------------------

* Load monthly Ibex-35 returns
* Note: In practice, load from tempfile created in data cleaning notebook
* For this notebook, we'll reload the data
* Run 00_data_cleaning.ipynb first to create tempfiles

* Load IBEX-35 monthly data
* (Assuming data cleaning notebook has been run)
* use `ibex35_monthly', clear

* For standalone execution, reload from raw data:
import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ibex35_close
keep date ibex35_close
drop if missing(ibex35_close)
sort date
duplicates drop date, force
tsset date
gen year = year(date)
gen month = month(date)
gen date_m = ym(year, month)
format date_m %tm
bysort date_m: egen temp_last = max(date)
keep if date == temp_last
keep date_m ibex35_close
rename ibex35_close ibex35_close_m
tsset date_m
gen log_ibex_m = ln(ibex35_close_m)
gen ribex35_monthly = 100 * D.log_ibex_m
drop log_ibex_m
label var ribex35_monthly "Ibex-35 Monthly Returns (%)"

* Keep relevant period
keep if date >= ym(1990, 2) & date <= ym(2014, 8)

* Time series plot
tsline ribex35_monthly, ///
    title("IBEX-35 Monthly Returns") ///
    ytitle("Returns (%)") xtitle("Date") ///
    name(ibex_returns_ts, replace)

* Autocorrelation
ac ribex35_monthly, lags(19) ///
    title("Autocorrelation: IBEX-35 Returns") ///
    name(acf_ibex, replace)

display "IBEX-35 Returns Summary Statistics:"
summarize ribex35_monthly, detail
```

**Interpretation**: The ACF plot shows autocorrelation coefficients at lags 1-19. For stock returns, we typically expect:
- **Low autocorrelation** at all lags (consistent with weak-form market efficiency)
- Most coefficients should fall within the confidence bands (not statistically significant)
- This suggests that past returns have little predictive power for future returns

**Economic Implication**: If stock returns were predictable from past returns, investors could exploit this to earn excess returns, which would violate the efficient market hypothesis (at least in its weak form).

<h2 id="section-5" class="styled-header">5) Exercise 4 - Autocorrelation with Confidence Intervals</h2>

**Slide Reference**: 12/69

We extend the autocorrelation analysis by examining more lags (up to 40) and using **Bartlett's confidence bands** for statistical inference.

**Bartlett's Confidence Bands**: For a white noise process, the 95% confidence interval for the sample ACF at lag $h$ is approximately:

$$\hat{\rho}(h) \pm 1.96 \times \frac{1}{\sqrt{T}}$$

If autocorrelation coefficients fall outside these bands, we can reject the null hypothesis that $\rho(h) = 0$ at the 5% significance level.


```python
* ----------------------------------------------------------------------------
* EXERCISE 4: Autocorrelation with confidence intervals
* Slide 12/69
* ----------------------------------------------------------------------------

* IBEX-35 returns
* Reload monthly IBEX data
import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ibex35_close
keep date ibex35_close
drop if missing(ibex35_close)
sort date
duplicates drop date, force
tsset date
gen year = year(date)
gen month = month(date)
gen date_m = ym(year, month)
format date_m %tm
bysort date_m: egen temp_last = max(date)
keep if date == temp_last
keep date_m ibex35_close
rename ibex35_close ibex35_close_m
tsset date_m
gen log_ibex_m = ln(ibex35_close_m)
gen ribex35_monthly = 100 * D.log_ibex_m
drop log_ibex_m
label var ribex35_monthly "Ibex-35 Monthly Returns (%)"

keep if date >= ym(1990, 2) & date <= ym(2014, 8)

ac ribex35_monthly, lags(40) ///
    title("ACF: IBEX-35 Returns with Bartlett's 95% Confidence Bands") ///
    name(acf_ibex_40, replace)
```

**Interpretation**:
- The **gray bands** represent Bartlett's 95% confidence intervals
- Most autocorrelation coefficients fall **within the confidence bands**, indicating they are **not statistically significant**
- This is consistent with the **weak-form efficient market hypothesis**: past returns do not predict future returns
- Any coefficients outside the bands may be due to **sampling variation** rather than true autocorrelation

**Statistical Test**: If we observe many coefficients outside the confidence bands, we would reject the null hypothesis that the series is white noise. For IBEX-35 returns, we typically find few significant autocorrelations, supporting market efficiency.
