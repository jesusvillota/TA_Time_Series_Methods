<h2 class="styled-header">1) Introduction to Seasonality</h2>

**Seasonality** refers to periodic patterns that repeat at regular intervals (e.g., monthly, quarterly, or yearly cycles).

**Common Sources of Seasonality**:
- **Weather**: Agricultural production, energy consumption
- **Holidays**: Retail sales, travel
- **Fiscal/Administrative**: Tax collection, government spending
- **Institutional**: School calendars, business cycles

**Detecting Seasonality**:
- Visual inspection of time series plots
- ACF analysis: Look for spikes at seasonal lags (e.g., lag 12 for monthly data)
- Seasonal unit root tests
- Decomposition methods

**Handling Seasonality**:
- **Seasonal differencing**: $\Delta_s Y_t = Y_t - Y_{t-s}$ where $s$ is the seasonal period
- **Seasonal ARIMA**: ARIMA models with seasonal components
- **Seasonal adjustment**: Remove seasonal component before modeling

<h2 class="styled-header">2) Exercise 6 - Spanish Inflation Seasonality</h2>

**Slide Reference**: 22/69

We analyze seasonal patterns in Spanish inflation by comparing:
1. **Monthly inflation rates**: May show strong seasonal patterns
2. **Annual inflation rates**: Should remove most seasonal effects

**Expected Patterns**:
- **Monthly inflation**: Strong autocorrelation at lag 12 (yearly cycle)
- **Annual inflation**: Reduced seasonal autocorrelation

### 2.1 Monthly vs Annual Inflation Rates

```stata
* Load Spanish CPI data from processed file
global processed_data "../../data/processed"
use "$processed_data/esp_cpi_monthly.dta", clear
```

**Visual Comparison**:
- **Monthly inflation**: Shows clear **seasonal fluctuations** (higher in some months, lower in others)
- **Annual inflation**: Shows **smoother** pattern with reduced seasonal variation

This suggests that taking **year-over-year differences** removes most seasonal effects.

### 2.2 ACF Analysis of Seasonal Patterns

**ACF Patterns**:
- **Monthly inflation**: Strong autocorrelation at **lag 12** (yearly cycle) - this is the seasonal pattern
- **Annual inflation**: Reduced autocorrelation at lag 12, suggesting seasonal effects are removed

**Interpretation**: The spike at lag 12 in monthly inflation indicates that inflation in January is correlated with inflation in the previous January, etc. This is the **seasonal component**.

### 2.3 Economic Interpretation

**Why Does Spanish Inflation Show Seasonality?**

1. **Food prices**: Seasonal patterns in agricultural production
2. **Energy prices**: Higher consumption in winter months
3. **Tourism**: Summer months see higher prices in tourist areas
4. **Holiday effects**: Christmas shopping affects December prices

**Policy Implications**:
- Central banks often focus on **year-over-year** inflation to avoid seasonal noise
- **Core inflation** (excluding food and energy) may show less seasonality
- Understanding seasonality helps distinguish **temporary** vs **persistent** inflation changes

**Modeling Implications**:
- For monthly data with seasonality, consider **SARIMA** models (Seasonal ARIMA)
- Or use **seasonally adjusted** data before applying standard ARMA models
- Annual differences ($Y_t - Y_{t-12}$) can remove seasonal effects
