<h2 id="section-1" class="styled-header">1) Introduction to Stationarity</h2>

**Stationarity** is a fundamental concept in time series analysis. A time series is **stationary** if its statistical properties (mean, variance, autocovariance) do not change over time.

**Why Stationarity Matters**:
- Standard statistical inference requires stationarity
- Non-stationary series can lead to **spurious regressions**
- Many time series models (ARMA) assume stationarity
- Non-stationary series often need transformation (differencing) before modeling

**Types of Stationarity**:

1. **Strict Stationarity**: Joint distribution invariant to time shifts
2. **Weak (Covariance) Stationarity**: 
   - Constant mean: $\mathbb{E}[Y_t] = \mu$ for all $t$
   - Constant variance: $\mathbb{V}[Y_t] = \sigma^2$ for all $t$
   - Autocovariance depends only on lag: $\gamma(t, t-h) = \gamma(h)$

**Unit Root Tests**: The **DF-GLS test** (Dickey-Fuller Generalized Least Squares) tests whether a series has a unit root (non-stationary). 

- **$H_0$**: Series has a unit root (non-stationary)
- **$H_1$**: Series is stationary

If we **reject** $H_0$, the series is stationary. If we **fail to reject**, the series is non-stationary and may need differencing.


<h2 id="section-2" class="styled-header">2) Exercise 5 - Unit Root Tests for IBEX-35</h2>

**Slide References**: 16/69, 17/69, 18/69, 19/69

We test for stationarity in IBEX-35 data by comparing:
1. **Index levels** (prices) - expected to be **non-stationary**
2. **Returns** (first differences of log prices) - expected to be **stationary**

**Expected Results**:
- **Levels**: Unit root test should **fail to reject** $H_0$ (non-stationary)
- **Returns**: Unit root test should **reject** $H_0$ (stationary)

This demonstrates that **differencing** transforms a non-stationary series into a stationary one.


### 2.1 Time Series Plots: Levels vs Returns


```stata
* ----------------------------------------------------------------------------
* EXERCISE 5: Stationarity - Dickey-Fuller Unit root test for Ibex-35
* Slides 16/69, 17/69, 18/69, 19/69
* ----------------------------------------------------------------------------

* Load IBEX-35 daily data
* Note: In practice, load from tempfile created in data cleaning notebook
global raw_data "../../data/raw"

import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ibex35_close
keep date ibex35_close
drop if missing(ibex35_close)
sort date
duplicates drop date, force
tsset date

* Create returns (log differences)
gen log_ibex = ln(ibex35_close)
gen ribex35 = 100 * D.log_ibex
drop log_ibex
label var ribex35 "Ibex-35 Daily Returns (%)"

keep if date >= td(01jan1990) & date <= td(01jan2014)

* Time series plots
tsline ibex35_close, ///
    title("IBEX-35 Index Levels") ///
    ytitle("Index") xtitle("Date") ///
    name(ibex_levels, replace)

tsline ribex35, ///
    title("IBEX-35 Daily Returns") ///
    ytitle("Returns (%)") xtitle("Date") ///
    name(ibex_returns, replace)
```

**Visual Inspection**:
- **Levels**: Show a clear **trend** (non-stationary - mean changes over time)
- **Returns**: Fluctuate around **zero** (stationary - constant mean)

This visual inspection suggests that levels are non-stationary while returns are stationary.


### 2.2 Correlograms: Levels vs Returns


```stata
* Correlograms
ac ibex35_close, lags(40) ///
    title("ACF: IBEX-35 Levels") ///
    name(acf_ibex_levels, replace)

ac ribex35, lags(40) ///
    title("ACF: IBEX-35 Returns") ///
    name(acf_ibex_ret, replace)
```

**ACF Patterns**:
- **Levels**: ACF decays **very slowly** (characteristic of non-stationary series)
- **Returns**: ACF decays **quickly** and stays within confidence bands (characteristic of stationary series)

The slow decay in levels indicates **long memory** and suggests a unit root process.


### 2.3 DF-GLS Unit Root Tests


```stata
* DF-GLS unit root tests
* Create continuous series for dfgls (requires no gaps)
preserve
drop if missing(ibex35_close)
gen t = _n
tsset t
display "DF-GLS Test for IBEX-35 Levels:"
quietly dfgls ibex35_close, maxlag(3) notrend
display "  Test statistic at lag 1: " r(dfgls1)
display "  Test statistic at lag 2: " r(dfgls2)
display "  Test statistic at lag 3: " r(dfgls3)
display "  Observations: " r(N)
display ""
display "  Interpretation: If test statistic > critical value, fail to reject H0 (non-stationary)"
display "                  If test statistic < critical value, reject H0 (stationary)"
restore

preserve
drop if missing(ribex35)
gen t = _n
tsset t
display "DF-GLS Test for IBEX-35 Returns:"
quietly dfgls ribex35, maxlag(3) notrend
display "  Test statistic at lag 1: " r(dfgls1)
display "  Test statistic at lag 2: " r(dfgls2)
display "  Test statistic at lag 3: " r(dfgls3)
display "  Observations: " r(N)
display ""
display "  Interpretation: Returns should reject H0 (stationary)"
restore
```

### 2.4 Interpretation and Economic Implications

**Test Results Interpretation**:

**For Levels (ibex35_close)**:
- Test statistics are typically **positive** or close to zero
- **Fail to reject** $H_0$: Series has a unit root (non-stationary)
- This confirms that **stock prices** follow a **random walk** or near-random walk process
- Consistent with the **efficient market hypothesis**: prices incorporate all available information

**For Returns (ribex35)**:
- Test statistics are typically **negative** and large in magnitude
- **Reject** $H_0$: Series is stationary
- This confirms that **returns** are stationary, making them suitable for ARMA modeling

**Economic Implications**:
1. **Price levels are unpredictable**: Cannot forecast future prices from past prices
2. **Returns are stationary**: Can model returns using ARMA/GARCH models
3. **Differencing transforms non-stationary to stationary**: $\Delta \ln(P_t) = r_t$ is stationary even though $\ln(P_t)$ is not
4. **Risk management**: Stationary returns allow for meaningful volatility modeling and risk measures

**Key Takeaway**: Always check for stationarity before applying time series models. If non-stationary, difference the series until it becomes stationary.
