<h2 id="section-1" class="styled-header">1) Introduction to ARMA Models</h2>

**ARMA(p,q)** models combine autoregressive (AR) and moving average (MA) components:

$$Y_t = \phi_1 Y_{t-1} + \cdots + \phi_p Y_{t-p} + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \cdots + \theta_q \varepsilon_{t-q}$$

where $\{\varepsilon_t\}$ is white noise.

**Model Selection**: Use information criteria:
- **AIC** (Akaike Information Criterion): $AIC = -2\ln(L) + 2k$
- **BIC** (Bayesian Information Criterion): $BIC = -2\ln(L) + k\ln(T)$

Lower values indicate better fit (penalizing for complexity).
