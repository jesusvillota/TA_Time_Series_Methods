<h2 class="styled-header">1) Introduction to VAR Models</h2>

**Vector Autoregression (VAR)** models extend univariate AR models to multiple time series.

**VAR(p) Model**:
$$\mathbf{Y}_t = \mathbf{c} + \mathbf{\Phi}_1 \mathbf{Y}_{t-1} + \cdots + \mathbf{\Phi}_p \mathbf{Y}_{t-p} + \boldsymbol{\varepsilon}_t$$

where $\mathbf{Y}_t$ is a vector of $k$ variables.

**Key Applications**:
- **Granger Causality**: Test if one variable helps predict another
- **Impulse Response Functions**: Trace dynamic effects of shocks
- **Variance Decomposition**: Quantify sources of forecast error variance

**Requirements**: All variables must be **stationary** (or cointegrated).
