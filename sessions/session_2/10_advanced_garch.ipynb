<h2 id="section-1" class="styled-header">1) Introduction to Advanced GARCH Models</h2>

This notebook covers advanced GARCH specifications:

1. **Asymmetric GARCH**: Captures leverage effect (negative returns increase volatility more)
2. **Non-normal distributions**: Student t innovations to account for fat tails
3. **Normality testing**: Jarque-Bera test for standardized residuals

**Asymmetric GARCH (TARCH/GJR-GARCH)**:
$$\sigma_t^2 = \omega + \alpha_1 \varepsilon_{t-1}^2 + \gamma_1 I_{t-1} \varepsilon_{t-1}^2 + \beta_1 \sigma_{t-1}^2$$

where $I_{t-1} = 1$ if $\varepsilon_{t-1} < 0$ (negative shock), and 0 otherwise.

**Student t GARCH**: Uses Student t distribution instead of normal for innovations, better capturing fat tails.

<h2 id="section-2" class="styled-header">2) Exercise 18 - Asymmetric GARCH(1,1)</h2>

**Slide Reference**: 29/41

We estimate an asymmetric GARCH model (TARCH/GJR-GARCH) to capture the leverage effect.

### 2.1 TARCH/GJR-GARCH Estimation

```stata
* Load IBEX-35 data
import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ibex35_close
keep date ibex35_close
drop if missing(ibex35_close)
sort date
duplicates drop date, force
tsset date
gen log_ibex = ln(ibex35_close)
gen ribex35 = 100 * D.log_ibex
keep if date >= td(01jan1990) & date <= td(01jan2015)

* Estimate Asymmetric GARCH (TARCH/GJR-GARCH) model
display "Asymmetric GARCH(1,1) Estimation:"
arch ribex35, arch(1) garch(1) tarch(1) vce(robust)
```

### 2.2 Leverage Effect

**Interpretation**: A positive coefficient on the TARCH term indicates the **leverage effect**: negative returns increase volatility more than positive returns of the same magnitude. This is consistent with financial theory where bad news (negative returns) increases uncertainty more than good news.

<h2 id="section-3" class="styled-header">3) Exercise 20 - Testing for Gaussianity</h2>

**Slide Reference**: 35/41

We test whether standardized residuals from GARCH(1,1) follow a normal distribution using the Jarque-Bera test.

### 3.1 Jarque-Bera Test

```stata
* Estimate GARCH(1,1) to get standardized residuals
quietly arch ribex35, arch(1) garch(1)
predict vol_garch, variance
predict resid_garch, residuals
gen std_resid = resid_garch / sqrt(vol_garch)

* Jarque-Bera test for normality
display "Jarque-Bera Test for Normality of Standardized Residuals:"
quietly summarize std_resid, detail
scalar skew = r(skewness)
scalar kurt = r(kurtosis)
scalar n = r(N)

* Calculate JB statistic
scalar jb = n/6 * (skew^2 + (kurt-3)^2/4)

display "  Skewness: " skew
display "  Kurtosis: " kurt
display "  JB statistic: " jb
display "  (Reject normality if JB is large)"
display ""
```

### 3.2 Normality Assessment

**Interpretation**:
- **Skewness**: Measures asymmetry (normal distribution has skewness = 0)
- **Kurtosis**: Measures tail thickness (normal distribution has kurtosis = 3)
- **Jarque-Bera statistic**: Tests joint hypothesis of zero skewness and kurtosis = 3
  - Under null hypothesis (normality), JB ~ $\chi^2(2)$
  - Large JB values → reject normality

Financial returns typically show **negative skewness** and **excess kurtosis** (kurtosis > 3), indicating fat tails and non-normality.

<h2 id="section-4" class="styled-header">4) Exercise 21 - GARCH with Student t Innovations</h2>

**Slide References**: 37/41-39/41

We estimate GARCH(1,1) with Student t distribution for innovations to better capture fat tails.

### 4.1 Student t GARCH Estimation

```stata
* Estimate GARCH(1,1) with Student t distribution
display "GARCH(1,1) with Student t innovations:"
arch ribex35, arch(1) garch(1) distribution(t) vce(robust)

* Extract degrees of freedom
matrix b = e(b)
display ""
display "Degrees of freedom: " b[1,5]
display ""
```

### 4.2 Distribution Comparison

```stata
* Compare kernel densities
predict resid_t, residuals
predict vol_t, variance
gen std_resid_t = resid_t / sqrt(vol_t)

kdensity std_resid_t, normal ///
    title("Standardized Residuals: Kernel vs Normal vs Student t") ///
    name(kde_garch_t, replace)
```

**Interpretation**:

- **Degrees of freedom**: Lower values indicate fatter tails (closer to normal as df → ∞)
- **Student t distribution**: Better captures the fat tails observed in financial returns
- **Model fit**: Student t GARCH typically provides better fit than normal GARCH for financial data

The kernel density plot shows how the Student t distribution better matches the empirical distribution of standardized residuals, especially in the tails.
