<h2 id="section-1" class="styled-header">1) Introduction to GARCH Models</h2>

**GARCH(p,q)** (Generalized Autoregressive Conditional Heteroskedasticity) models extend ARCH by allowing volatility to depend on both past squared innovations and past volatility:

$$\sigma_t^2 = \omega + \sum_{i=1}^q \alpha_i \varepsilon_{t-i}^2 + \sum_{j=1}^p \beta_j \sigma_{t-j}^2$$

**GARCH(1,1)** is the most common specification:

$$\sigma_t^2 = \omega + \alpha_1 \varepsilon_{t-1}^2 + \beta_1 \sigma_{t-1}^2$$

**Key Advantages over ARCH**:
- **More parsimonious**: Fewer parameters than ARCH(q) for similar fit
- **Captures long memory**: Volatility persistence through $\beta_1$
- **Persistence**: $\alpha_1 + \beta_1$ measures how long shocks persist

**Persistence Interpretation**:
- If $\alpha_1 + \beta_1 < 1$: Stationary volatility (mean-reverting)
- If $\alpha_1 + \beta_1 \approx 1$: Near-integrated (IGARCH)
- **Half-life**: $-\ln(2)/\ln(\alpha_1 + \beta_1)$ days

<h2 id="section-2" class="styled-header">2) Exercise 16 - ARCH vs GARCH Comparison</h2>

**Slide Reference**: 27/41

We compare model fit across ARCH and GARCH specifications to demonstrate the parsimony advantage of GARCH models.


### 2.1 Model Comparison Table


```stata
* Load IBEX-35 data
import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ibex35_close
keep date ibex35_close
drop if missing(ibex35_close)
sort date
duplicates drop date, force
tsset date
gen log_ibex = ln(ibex35_close)
gen ribex35 = 100 * D.log_ibex
keep if date >= td(01jan1990) & date <= td(01jan2015)

* Estimate models and store results
display "Model Comparison:"
display ""

quietly arch ribex35, arch(1)
scalar ll_arch1 = e(ll)
scalar k_arch1 = e(k)

quietly arch ribex35, arch(1/10)
scalar ll_arch10 = e(ll)
scalar k_arch10 = e(k)

quietly arch ribex35, arch(1) garch(1)
scalar ll_garch11 = e(ll)
scalar k_garch11 = e(k)

display "Model         Log-Likelihood    Parameters"
display "ARCH(1)       " ll_arch1 "           " k_arch1
display "ARCH(10)      " ll_arch10 "          " k_arch10
display "GARCH(1,1)    " ll_garch11 "          " k_garch11
```

### 2.2 Parsimony of GARCH

**Interpretation**: GARCH(1,1) achieves similar or better fit than ARCH(10) with fewer parameters. This demonstrates the **parsimony advantage** of GARCH models:

- **ARCH(1)**: 2 parameters (mean + ARCH(1))
- **ARCH(10)**: 11 parameters (mean + ARCH(1) through ARCH(10))
- **GARCH(1,1)**: 3 parameters (mean + ARCH(1) + GARCH(1))

GARCH(1,1) captures long memory in volatility with just one additional parameter compared to ARCH(1).


<h2 id="section-3" class="styled-header">3) Exercise 17 - GARCH(1,1) Model</h2>

**Slide References**: 23/41-26/41

We estimate a GARCH(1,1) model for IBEX-35 returns and analyze volatility persistence and forecasting.


### 3.1 GARCH(1,1) Estimation


```stata
* Estimate GARCH(1,1) model
display "GARCH(1,1) Estimation:"
arch ribex35, arch(1) garch(1) vce(robust)
```

### 3.2 Persistence and Half-Life


```stata
* Store coefficient estimates
matrix b = e(b)
scalar alpha0 = b[1,2]
scalar alpha1 = b[1,3]
scalar beta1 = b[1,4]

display ""
display "Persistence: alpha1 + beta1 = " alpha1 + beta1
display "Half-life of shocks (approx): " -ln(2)/ln(alpha1 + beta1) " days"
display ""
```

**Interpretation**:
- **Persistence** ($\alpha_1 + \beta_1$): Measures how long volatility shocks persist
  - Values close to 1 indicate high persistence (shocks decay slowly)
  - Values < 1 indicate mean-reverting volatility
- **Half-life**: Number of days for a volatility shock to decay to half its initial value
  - Higher persistence → longer half-life


### 3.3 Volatility Forecasting


```stata
* Predict conditional volatility
predict vol_garch11, variance
gen log_vol_garch11 = ln(vol_garch11)

* Plot volatility
tsline log_vol_garch11, ///
    title("GARCH(1,1): Log Volatility Estimates") ///
    ytitle("Log Volatility") xtitle("Date") ///
    name(vol_garch11_plot, replace)
```

### 3.4 Residual Diagnostics


```stata
* Standardized residuals
predict resid_garch11, residuals
gen std_resid_garch11 = resid_garch11 / sqrt(vol_garch11)
gen std_resid_sq_garch11 = std_resid_garch11^2

* Diagnostics
ac std_resid_sq_garch11, lags(40) ///
    title("ACF: Squared Standardized Residuals (GARCH(1,1))") ///
    name(acf_std_resid_garch11, replace)

pac std_resid_sq_garch11, lags(40) ///
    title("PACF: Squared Standardized Residuals (GARCH(1,1))") ///
    name(pacf_std_resid_garch11, replace)
```

**Interpretation**: If the GARCH(1,1) model adequately captures volatility dynamics, the squared standardized residuals should show no significant autocorrelation. This indicates that the model has successfully removed volatility clustering from the residuals.


<h2 id="section-4" class="styled-header">4) Exercise 19 - Extended Model Comparison</h2>

**Slide Reference**: 30/41

We extend the model comparison to include asymmetric GARCH specifications that capture the leverage effect.


### 4.1 Including Asymmetric GARCH


```stata
* Add asymmetric GARCH to comparison
quietly arch ribex35, arch(1) garch(1) tarch(1)
scalar ll_tgarch11 = e(ll)
scalar k_tgarch11 = e(k)

display "Model              Log-Likelihood    Parameters"
display "ARCH(1)            " ll_arch1 "           " k_arch1
display "ARCH(10)           " ll_arch10 "          " k_arch10
display "GARCH(1,1)         " ll_garch11 "          " k_garch11
display "Asymm GARCH(1,1)   " ll_tgarch11 "          " k_tgarch11
display ""
```

### 4.2 Model Selection

**Interpretation**: The extended comparison shows:

1. **GARCH(1,1) vs ARCH models**: GARCH provides better fit with fewer parameters
2. **Asymmetric GARCH**: May improve fit if leverage effect is present
3. **Model selection**: Use information criteria (AIC/BIC) or likelihood ratio tests to choose between symmetric and asymmetric specifications

**Trade-off**: Asymmetric GARCH adds one parameter but may better capture the leverage effect observed in financial returns.
