<h2 id="section-1" class="styled-header">1) Introduction to Volatility Modeling</h2>

Financial returns exhibit several **stylized facts** that motivate volatility modeling:

1. **Returns show little autocorrelation** (weak-form market efficiency)
2. **Squared returns show strong persistence** (volatility clustering)
3. **Leverage effect**: Negative returns increase volatility more than positive returns
4. **Non-normality**: Returns have fat tails and excess kurtosis

These facts suggest that **volatility is time-varying** and requires specialized models (ARCH/GARCH).
