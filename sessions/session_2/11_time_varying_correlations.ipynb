<h2 id="section-1" class="styled-header">1) Introduction to Time-Varying Correlations</h2>

Correlations between financial assets are **not constant** over time. This notebook analyzes:

1. **Rolling correlations**: Using a moving window (e.g., 100 days) to estimate time-varying correlations
2. **Cross-asset correlations**: Between stock indices and sovereign bonds
3. **Correlation patterns**: How correlations change during crises (contagion effects)

**Key Finding**: Correlations tend to **increase during crises**, reducing diversification benefits when they are most needed.

```stata
* Setup
global raw_data "../../../data/raw"
```

<h2 id="section-2" class="styled-header">2) Exercise 22 - Correlations Between Euro Area Stock Indices</h2>

**Slide Reference**: 7/33

We calculate rolling 100-day correlations between returns of euro area reference stock indices.

### 2.1 Rolling Correlation Calculation

```stata
* Load all euro area stock indices
import delimited "$raw_data/IBEX35.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ibex35_close
keep date ibex35_close
drop if missing(ibex35_close)
sort date
duplicates drop date, force
tsset date
gen log_ibex = ln(ibex35_close)
gen ribex35 = 100 * D.log_ibex
keep date ribex35
tempfile ibex35
save `ibex35'

* CAC 40
import delimited "$raw_data/CAC40.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close cac40_close
keep date cac40_close
drop if missing(cac40_close)
sort date
duplicates drop date, force
tsset date
gen log_cac = ln(cac40_close)
gen rcac40 = 100 * D.log_cac
keep date rcac40
tempfile cac40
save `cac40'

* DAX
import delimited "$raw_data/DAX.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close dax_close
keep date dax_close
drop if missing(dax_close)
sort date
duplicates drop date, force
tsset date
gen log_dax = ln(dax_close)
gen rdax = 100 * D.log_dax
keep date rdax
tempfile dax
save `dax'

* FTSE MIB
import delimited "$raw_data/FTSEMIB.csv", clear varnames(1) rowrange(4) stringcols(_all)
gen date_numeric = date(price, "YMD")
format date_numeric %td
rename date_numeric date
destring close, replace force
rename close ftsemib_close
keep date ftsemib_close
drop if missing(ftsemib_close)
sort date
duplicates drop date, force
tsset date
gen log_ftse = ln(ftsemib_close)
gen rftsemib = 100 * D.log_ftse
keep date rftsemib
tempfile ftsemib
save `ftsemib'

* Merge all indices
use `ibex35', clear
merge 1:1 date using `cac40', nogen keep(match)
merge 1:1 date using `dax', nogen keep(match)
merge 1:1 date using `ftsemib', nogen keep(match)

* Keep period May 1998 to May 2014
keep if date >= td(01may1998) & date <= td(01may2014)
tsset date

* Calculate rolling 100-day correlations
gen corr_es_fr = .
gen corr_es_de = .
gen corr_es_it = .
gen corr_fr_de = .
gen corr_fr_it = .
gen corr_de_it = .

local window = 100
forvalues i = `window'/`=_N' {
    quietly correlate ribex35 rcac40 in `=`i'-`window'+1'/`i'
    quietly replace corr_es_fr = r(rho) in `i'
    
    quietly correlate ribex35 rdax in `=`i'-`window'+1'/`i'
    quietly replace corr_es_de = r(rho) in `i'
    
    quietly correlate ribex35 rftsemib in `=`i'-`window'+1'/`i'
    quietly replace corr_es_it = r(rho) in `i'
    
    quietly correlate rcac40 rdax in `=`i'-`window'+1'/`i'
    quietly replace corr_fr_de = r(rho) in `i'
    
    quietly correlate rcac40 rftsemib in `=`i'-`window'+1'/`i'
    quietly replace corr_fr_it = r(rho) in `i'
    
    quietly correlate rdax rftsemib in `=`i'-`window'+1'/`i'
    quietly replace corr_de_it = r(rho) in `i'
}

* Calculate min, median, max across all correlations
egen corr_min = rowmin(corr_es_fr corr_es_de corr_es_it corr_fr_de corr_fr_it corr_de_it)
egen corr_median = rowmedian(corr_es_fr corr_es_de corr_es_it corr_fr_de corr_fr_it corr_de_it)
egen corr_max = rowmax(corr_es_fr corr_es_de corr_es_it corr_fr_de corr_fr_it corr_de_it)
```

### 2.2 Time-Varying Correlation Patterns

```stata
* Plot time-varying correlations
tsline corr_min corr_median corr_max, ///
    title("Time-Varying Correlations: Euro Area Stock Indices") ///
    ytitle("Correlation") xtitle("Date") ///
    legend(label(1 "Minimum") label(2 "Median") label(3 "Maximum")) ///
    name(corr_euro_indices, replace)
```

**Interpretation**:

- **Time-varying correlations**: Correlations change substantially over time
- **Crisis periods**: Correlations increase during financial crises (e.g., 2008-2009, 2011-2012)
- **Contagion effect**: Higher correlations during crises indicate that diversification benefits are reduced when most needed
- **Range**: The spread between min and max correlations shows the degree of time variation

<h2 id="section-3" class="styled-header">3) Exercise 23 - Stock Indices and Sovereign Bonds</h2>

**Slide Reference**: 8/33

We analyze correlations between stock index returns and sovereign bond returns using rolling windows.

### 3.1 Monthly Returns and Correlations

```stata
* Convert stock returns to monthly
gen year = year(date)
gen month = month(date)
gen date_m = ym(year, month)
format date_m %tm

* Aggregate to monthly: sum daily returns
collapse (sum) ribex35 rcac40 rdax, by(date_m)
rename date_m date

* Load bond data
import delimited "$raw_data/BOND_10Y_DE.csv", clear
gen date_numeric = date(observation_date, "YMD")
format date_numeric %td
gen year = year(date_numeric)
gen month = month(date_numeric)
gen date_m = ym(year, month)
format date_m %tm
drop observation_date date_numeric year month
rename date_m date
rename irltlt01dem156n bond_de
destring bond_de, replace force
sort date
duplicates drop date, force
tsset date
gen rbond_de = D.bond_de
tempfile bond_de
save `bond_de'

* Spain bonds
import delimited "$raw_data/BOND_10Y_ES.csv", clear
gen date_numeric = date(observation_date, "YMD")
format date_numeric %td
gen year = year(date_numeric)
gen month = month(date_numeric)
gen date_m = ym(year, month)
format date_m %tm
drop observation_date date_numeric year month
rename date_m date
rename irltlt01esm156n bond_es
destring bond_es, replace force
sort date
duplicates drop date, force
tsset date
gen rbond_es = D.bond_es
tempfile bond_es
save `bond_es'

* France bonds
import delimited "$raw_data/BOND_10Y_FR.csv", clear
gen date_numeric = date(observation_date, "YMD")
format date_numeric %td
gen year = year(date_numeric)
gen month = month(date_numeric)
gen date_m = ym(year, month)
format date_m %tm
drop observation_date date_numeric year month
rename date_m date
rename irltlt01frm156n bond_fr
destring bond_fr, replace force
sort date
duplicates drop date, force
tsset date
gen rbond_fr = D.bond_fr
tempfile bond_fr
save `bond_fr'

* Merge stock and bond data
preserve
use `ibex35', clear
gen year = year(date)
gen month = month(date)
gen date_m = ym(year, month)
format date_m %tm
collapse (sum) ribex35, by(date_m)
rename date_m date
tempfile stocks
save `stocks'
restore

use `stocks', clear
merge 1:1 date using `bond_es', nogen keep(match)
merge 1:1 date using `bond_fr', nogen keep(match)
merge 1:1 date using `bond_de', nogen keep(match)

* Load CAC and DAX monthly
preserve
use `cac40', clear
gen year = year(date)
gen month = month(date)
gen date_m = ym(year, month)
format date_m %tm
collapse (sum) rcac40, by(date_m)
rename date_m date
tempfile cac_monthly
save `cac_monthly'
restore

preserve
use `dax', clear
gen year = year(date)
gen month = month(date)
gen date_m = ym(year, month)
format date_m %tm
collapse (sum) rdax, by(date_m)
rename date_m date
tempfile dax_monthly
save `dax_monthly'
restore

merge 1:1 date using `cac_monthly', nogen keep(match)
merge 1:1 date using `dax_monthly', nogen keep(match)

* Keep period Aug 1991 to Aug 2015
keep if date >= ym(1991, 8) & date <= ym(2015, 8)
tsset date

* Calculate rolling 100-month correlations
gen corr_es_stock_bond = .
gen corr_fr_stock_bond = .
gen corr_de_stock_bond = .

local window = 100
forvalues i = `window'/`=_N' {
    quietly correlate ribex35 rbond_es in `=`i'-`window'+1'/`i'
    quietly replace corr_es_stock_bond = r(rho) in `i'
    
    quietly correlate rcac40 rbond_fr in `=`i'-`window'+1'/`i'
    quietly replace corr_fr_stock_bond = r(rho) in `i'
    
    quietly correlate rdax rbond_de in `=`i'-`window'+1'/`i'
    quietly replace corr_de_stock_bond = r(rho) in `i'
}
```

### 3.2 Cross-Asset Correlations

```stata
* Plot correlations
tsline corr_es_stock_bond corr_fr_stock_bond corr_de_stock_bond, ///
    title("Correlations: Stock Index vs Sovereign Bond Returns") ///
    ytitle("Correlation") xtitle("Date") ///
    legend(label(1 "Spain") label(2 "France") label(3 "Germany")) ///
    name(corr_stock_bond, replace)
```

**Interpretation**:

- **Cross-asset correlations**: Stock-bond correlations vary over time and can change sign
- **Flight to quality**: During crises, correlations may become negative (stocks fall, bonds rise)
- **Country differences**: Different patterns across countries reflect different risk perceptions
- **Portfolio implications**: Time-varying correlations affect optimal portfolio allocation
